```python
# ==============================================================================
# MODEL 1 (EARNINGS DAY) - PER-BANK FULL PIPELINE
# FIXED:
# - If BEST OVERALL MODEL is VAR, use TRUE VAR-based feature importance
#   from fitted VAR coefficients for the PCL_t+1 equation
# - ARIMAX / VARMAX use model-based coefficients where available
# - ML models still use built-in / coefficient importance
# - XGBoost uses built-in feature importance instead of SHAP for stability
# - Best overall model for forecasting stays LOWEST OOS MAPE
# - Best ML model for ML-specific importance uses LOWEST OOS RMSE
# ==============================================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.linear_model import ElasticNetCV, RidgeCV, LassoCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.inspection import permutation_importance
from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import TimeSeriesSplit

# Optional packages
try:
    from pmdarima import auto_arima
    HAS_PMDARIMA = True
except ImportError:
    HAS_PMDARIMA = False

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

try:
    from lightgbm import LGBMRegressor
    HAS_LGBM = True
except ImportError:
    HAS_LGBM = False

try:
    import shap
    HAS_SHAP = True
except ImportError:
    HAS_SHAP = False

PD_VER = int(pd.__version__.split(".")[0])
Q_FREQ = "QE" if PD_VER >= 2 else "Q"

print(">>> Building feature pool for Model 1 (Per-Bank Earnings Day Forecasts)...")
if not HAS_XGB or not HAS_LGBM:
    print("⚠️ Notice: xgboost or lightgbm not found. Run '!pip install xgboost lightgbm' if needed.")
if not HAS_SHAP:
    print("⚠️ Notice: shap not found. Tree-model SHAP charts will be skipped if unavailable.")

# ==============================================================================
# HELPERS
# ==============================================================================

def clean_num(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)

    s = str(x).strip()

    if s in ["—", "", "-", "NA", "N/A", "nan", "NaN", "None", "null", "NULL",
             "Unnamed:23", "Unnamed:39"]:
        return np.nan

    if s.startswith("(") and s.endswith(")"):
        s = "-" + s[1:-1]

    s = (
        s.replace(",", "")
         .replace("$", "")
         .replace(" ", "")
         .replace("[", "")
         .replace("]", "")
         .replace("{", "")
         .replace("}", "")
         .replace("(", "")
         .replace(")", "")
         .replace("'", "")
         .replace('"', "")
         .replace("\u200b", "")
         .replace("\xa0", "")
    )

    try:
        return float(s)
    except:
        return np.nan

def safe_mape(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    denom = np.where(np.abs(y_true) < 1e-8, np.nan, np.abs(y_true))
    return np.nanmean(np.abs((y_true - y_pred) / denom)) * 100

def plot_feature_bar(imp_df, bank, model_name, value_col="Feature Importance", xlabel="Importance"):
    if imp_df.empty:
        print(f"⚠️ No feature importance available for {bank}.")
        return

    imp_df_plot = imp_df.sort_values(value_col, ascending=True)

    plt.figure(figsize=(8, max(4, len(imp_df_plot) * 0.6)))
    plt.barh(imp_df_plot["Feature"], imp_df_plot[value_col])
    plt.xlabel(xlabel)
    plt.ylabel("Feature")
    plt.title(f"Feature Importance - {bank}\n{model_name}")
    plt.tight_layout()
    plt.show()

def quarter_str_from_date(dt):
    return f"{dt.year}Q{dt.quarter}"

def build_future_exog_df(bank_df, selected_features_m1, horizon=4):
    future_dates = pd.date_range(
        start=bank_df.index.max() + pd.offsets.QuarterEnd(1),
        periods=horizon,
        freq="Q"
    )

    feat_cols = [c for c in selected_features_m1 if c in bank_df.columns]
    future_exog = pd.DataFrame(index=future_dates, columns=feat_cols, dtype=float)

    for c in feat_cols:
        if c in bank_df.columns:
            for dt in future_dates:
                if dt in bank_df.index and pd.notna(bank_df.loc[dt, c]):
                    future_exog.loc[dt, c] = bank_df.loc[dt, c]
                else:
                    hist = pd.to_numeric(bank_df[c], errors="coerce").dropna()
                    future_exog.loc[dt, c] = hist.iloc[-1] if len(hist) > 0 else np.nan

    future_exog = future_exog.apply(pd.to_numeric, errors="coerce").ffill().bfill()
    return future_exog

def make_ml_models_dict(tscv):
    ml_models_dict = {
        "06. ML: Lasso": LassoCV(alphas=np.logspace(-2, 4, 50), cv=tscv),
        "07. ML: Ridge": RidgeCV(alphas=np.logspace(-2, 4, 50)),
        "08. ML: Random Forest": RandomForestRegressor(n_estimators=100, max_depth=3, random_state=42),
        "09. ML: Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
        "10. ML: Elastic Net": ElasticNetCV(l1_ratio=[0.1, 0.5, 0.9], cv=tscv, random_state=42)
    }

    if HAS_XGB:
        ml_models_dict["11. ML: XGBoost"] = XGBRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42
        )

    if HAS_LGBM:
        ml_models_dict["12. ML: LightGBM"] = LGBMRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42,
            verbose=-1
        )
    return ml_models_dict

def is_time_series_model(model_name):
    if pd.isna(model_name):
        return False
    model_name = str(model_name)
    return (
        model_name.startswith("01. Baseline") or
        model_name.startswith("02. TS:") or
        model_name.startswith("03. TS:") or
        model_name.startswith("04. TS:") or
        model_name.startswith("05. TS:")
    )

def var_feature_importance(train_valid_df1, features_ts_hist, target_col="PCL_t+1", maxlags=1, include_self_lag=False):
    """
    True VAR-based importance for the target equation using fitted VAR coefficients.
    Importance = sum of absolute lag coefficients for each variable in the target equation.
    By default excludes the target's own lag so chart focuses on explanatory variables.
    """
    var_cols = [target_col] + [c for c in features_ts_hist if c in train_valid_df1.columns]
    if len(var_cols) < 2:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    tmp = train_valid_df1[var_cols].apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan).dropna()
    if len(tmp) <= maxlags + 2:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    try:
        var_model = VAR(tmp).fit(maxlags=maxlags)
        params = var_model.params.copy()

        if target_col not in params.columns:
            return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

        features_for_importance = list(tmp.columns) if include_self_lag else [c for c in tmp.columns if c != target_col]

        out = []
        for feat in features_for_importance:
            lag_rows = [idx for idx in params.index if idx.startswith("L") and idx.endswith(f".{feat}")]
            abs_sum = params.loc[lag_rows, target_col].abs().sum() if len(lag_rows) > 0 else 0.0
            out.append({
                "Feature": feat,
                "Feature Importance": float(abs_sum),
                "Method": f"Abs VAR Coef Sum ({target_col} eqn)"
            })

        return pd.DataFrame(out).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

    except Exception as e:
        print(f"⚠️ VAR feature importance failed: {e}")
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

def arimax_feature_importance(train_valid_df1, exog_m1, target_col="PCL_t+1", order=(1, 0, 0)):
    """
    ARIMAX-based importance from absolute exogenous coefficients.
    """
    needed_cols = [target_col] + [c for c in exog_m1 if c in train_valid_df1.columns]
    if len(needed_cols) < 2:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    tmp = train_valid_df1[needed_cols].apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan).dropna()
    if len(tmp) < 6:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    try:
        model = ARIMA(
            tmp[target_col].values,
            exog=tmp[exog_m1].values,
            order=order
        ).fit()

        exog_coef = model.params[-len(exog_m1):]
        out = pd.DataFrame({
            "Feature": exog_m1,
            "Feature Importance": np.abs(exog_coef),
            "Coefficient": exog_coef,
            "Method": "Absolute ARIMAX Coefficient"
        }).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

        return out
    except Exception as e:
        print(f"⚠️ ARIMAX feature importance failed: {e}")
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

def varmax_feature_importance(train_valid_df1, features_ts_hist, exog_m1, target_col="PCL_t+1"):
    """
    VARMAX-based importance from absolute exogenous coefficients when available.
    Endogenous lag structure in VARMAX is harder to summarize cleanly, so this focuses on exog terms.
    """
    vmax_endog_cols = [target_col] + [c for c in features_ts_hist if c in train_valid_df1.columns]
    if len(vmax_endog_cols) < 2 or len(exog_m1) == 0:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    vmax_endog = train_valid_df1[vmax_endog_cols].apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan).dropna().copy()
    if len(vmax_endog) < 6:
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

    try:
        vmax_exog = train_valid_df1.loc[vmax_endog.index, exog_m1].apply(pd.to_numeric, errors="coerce")
        valid_idx = vmax_endog.index.intersection(vmax_exog.dropna().index)
        vmax_endog = vmax_endog.loc[valid_idx]
        vmax_exog = vmax_exog.loc[valid_idx]

        if len(vmax_endog) < 6:
            return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

        model = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)
        param_series = pd.Series(model.params, index=model.param_names)

        out = []
        for feat in exog_m1:
            matched = [name for name in param_series.index if (feat in name and "beta" in name.lower())]
            abs_sum = param_series.loc[matched].abs().sum() if len(matched) > 0 else 0.0
            out.append({
                "Feature": feat,
                "Feature Importance": float(abs_sum),
                "Method": f"Abs VARMAX Exog Coef Sum ({target_col} eqn)"
            })

        return pd.DataFrame(out).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

    except Exception as e:
        print(f"⚠️ VARMAX feature importance failed: {e}")
        return pd.DataFrame(columns=["Feature", "Feature Importance", "Method"])

def forecast_best_model_1y(
    bank,
    best_model_name,
    train_valid_df1,
    bank_df_full,
    selected_features_m1,
    exog_m1,
    features_ts_hist,
    horizon=4
):
    last_date = bank_df_full.index.max()
    future_dates = pd.date_range(
        start=last_date + pd.offsets.QuarterEnd(1),
        periods=horizon,
        freq="Q"
    )

    future_feat_df = build_future_exog_df(bank_df_full, selected_features_m1, horizon=horizon)
    full_target = pd.to_numeric(train_valid_df1["PCL_t+1"], errors="coerce").dropna().copy()

    def output_df(preds):
        return pd.DataFrame({
            "Bank": bank,
            "Forecast Quarter": [quarter_str_from_date(d) for d in future_dates],
            "Best Model": best_model_name,
            "Forecast ($M)": np.array(preds, dtype=float)
        })

    if best_model_name == "01. Baseline: Naive (Last Qtr)":
        if "PCL_t" in bank_df_full.columns and pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().shape[0] > 0:
            last_val = pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().iloc[-1]
        elif len(full_target) > 0:
            last_val = full_target.iloc[-1]
        else:
            last_val = np.nan
        return output_df([last_val] * horizon)

    elif best_model_name == "02. TS: SARIMA (Blind)":
        try:
            y = full_target.astype(float)
            if HAS_PMDARIMA:
                sarima_model = auto_arima(y, seasonal=False, suppress_warnings=True, error_action="ignore")
                preds = sarima_model.predict(n_periods=horizon)
            else:
                arima_model = ARIMA(y.values, order=(1, 0, 0)).fit()
                preds = arima_model.forecast(steps=horizon)
            return output_df(preds)
        except:
            fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
            return output_df([fallback] * horizon)

    elif best_model_name == "03. TS: VAR (Blind)":
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            var_input = train_valid_df1[var_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(var_input) > 5 and var_input.shape[1] >= 2:
                model = VAR(var_input).fit(maxlags=1)
                preds = []
                current_window = var_input.values.copy()
                last_macro = var_input.iloc[-1, 1:].values if var_input.shape[1] > 1 else np.array([])

                for _ in range(horizon):
                    fc = model.forecast(current_window[-model.k_ar:], steps=1)[0]
                    if len(last_macro) > 0:
                        fc[1:] = last_macro
                    preds.append(fc[0])
                    current_window = np.vstack([current_window, fc])

                return output_df(preds)
            else:
                fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
                return output_df([fallback] * horizon)
        except:
            fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
            return output_df([fallback] * horizon)

    elif best_model_name == "04. TS: ARIMAX (Macro-Aware)":
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            train_arimax = train_valid_df1[needed_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(train_arimax) > 5 and len(exog_m1) > 0:
                future_exog = build_future_exog_df(bank_df_full, exog_m1, horizon=horizon)[exog_m1]
                future_exog = future_exog.apply(pd.to_numeric, errors="coerce")

                model = ARIMA(
                    train_arimax["PCL_t+1"].values,
                    exog=train_arimax[exog_m1].values,
                    order=(1, 0, 0)
                ).fit()

                preds = model.forecast(steps=horizon, exog=future_exog.values)
                return output_df(preds)
            else:
                fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
                return output_df([fallback] * horizon)
        except:
            fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
            return output_df([fallback] * horizon)

    elif best_model_name == "05. TS: VARMAX (Macro-Aware)":
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            vmax_endog = train_valid_df1[vmax_endog_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_valid_df1.loc[vmax_endog.index, exog_m1].apply(pd.to_numeric, errors="coerce").copy()
                valid_idx = vmax_endog.index.intersection(vmax_exog.dropna().index)
                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]

                future_exog = build_future_exog_df(bank_df_full, exog_m1, horizon=horizon)[exog_m1]
                future_exog = future_exog.apply(pd.to_numeric, errors="coerce")

                model = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)
                fc = model.forecast(steps=horizon, exog=future_exog)
                preds = fc.iloc[:, 0].values
                return output_df(preds)
            else:
                fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
                return output_df([fallback] * horizon)
        except:
            fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
            return output_df([fallback] * horizon)

    elif "ML:" in best_model_name:
        try:
            full_ml_df = train_valid_df1[selected_features_m1 + ["PCL_t+1"]].copy()
            full_ml_df = full_ml_df.apply(pd.to_numeric, errors="coerce").dropna()

            if len(full_ml_df) <= 5 or len(selected_features_m1) == 0:
                fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
                return output_df([fallback] * horizon)

            n_splits_fit = min(3, len(full_ml_df) - 1)
            if n_splits_fit < 2:
                n_splits_fit = 2

            tscv_fit = TimeSeriesSplit(n_splits=n_splits_fit)
            ml_models_dict = make_ml_models_dict(tscv_fit)
            model = ml_models_dict.get(best_model_name)

            X_train = full_ml_df[selected_features_m1].astype(float)
            y_train = full_ml_df["PCL_t+1"].astype(float).values

            scaler = StandardScaler()
            X_train_sc = scaler.fit_transform(X_train)

            fitted_model = model.fit(X_train_sc, y_train)

            preds = []
            recursive_future = future_feat_df.copy()

            if "PCL_t" in recursive_future.columns:
                if "PCL_t" in bank_df_full.columns and pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().shape[0] > 0:
                    last_pcl = pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().iloc[-1]
                elif len(full_target) > 0:
                    last_pcl = full_target.iloc[-1]
                else:
                    last_pcl = np.nan
            else:
                last_pcl = None

            for i in range(horizon):
                row = recursive_future.iloc[i:i+1].copy()
                row = row.apply(pd.to_numeric, errors="coerce")

                if "PCL_t" in row.columns:
                    row.loc[:, "PCL_t"] = last_pcl

                row = row[selected_features_m1].astype(float)
                X_sc = scaler.transform(row)
                pred = float(fitted_model.predict(X_sc)[0])
                preds.append(pred)

                if "PCL_t" in recursive_future.columns:
                    last_pcl = pred

            return output_df(preds)

        except Exception as e:
            print(f"⚠️ Forecasting failed for {bank} / {best_model_name}: {e}")
            fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
            return output_df([fallback] * horizon)

    fallback = full_target.iloc[-1] if len(full_target) > 0 else np.nan
    return output_df([fallback] * horizon)

# ==============================================================================
# LOAD DATA
# ==============================================================================

file_hist = "Complete_UC2_Forecast_Data.xlsx"
raw_df = pd.read_excel(file_hist, sheet_name="Forecast Data", header=0)
raw_df.columns = [str(c).strip() for c in raw_df.columns]

print("\nDetected columns:")
print(raw_df.columns.tolist())

required_base_cols = ["Bank", "Year", "Quarter"]
missing_base_cols = [c for c in required_base_cols if c not in raw_df.columns]
if missing_base_cols:
    raise ValueError(f"Missing required columns: {missing_base_cols}")

raw_df["Bank"] = raw_df["Bank"].astype(str).str.strip()
raw_df["Year"] = pd.to_numeric(raw_df["Year"], errors="coerce")
raw_df["Quarter"] = raw_df["Quarter"].astype(str).str.strip()

raw_df = raw_df.dropna(subset=["Bank", "Year", "Quarter"])
raw_df = raw_df.sort_values(by=["Bank", "Year", "Quarter"]).reset_index(drop=True)

# CLEAN ALL NON-ID COLUMNS
id_cols = {"Bank", "Quarter"}
for c in raw_df.columns:
    if c not in id_cols:
        raw_df[c] = raw_df[c].apply(clean_num)
        raw_df[c] = pd.to_numeric(raw_df[c], errors="coerce")

remaining_object_cols = raw_df.select_dtypes(include=["object"]).columns.tolist()
print("\nRemaining object columns after cleaning:", remaining_object_cols)

cols_to_fill = [
    "PCL_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

for c in cols_to_fill:
    if c in raw_df.columns:
        raw_df[c] = raw_df.groupby("Bank")[c].transform(lambda x: x.ffill().bfill())

q_map = {"Q1": "03-31", "Q2": "06-30", "Q3": "09-30", "Q4": "12-31"}

candidate_features_m1 = [
    "PCL_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

features_ts_hist = ["GDP_YoY_t", "Overnight_Rate_t"]

# ==============================================================================
# BUILD PER-BANK DATASETS
# ==============================================================================

bank_model_data = {}

for bank in sorted(raw_df["Bank"].dropna().unique()):
    bank_df = raw_df[raw_df["Bank"] == bank].copy()

    bank_df["date_idx"] = pd.to_datetime(
        bank_df["Year"].astype("Int64").astype(str) + "-" + bank_df["Quarter"].map(q_map),
        errors="coerce"
    )

    bank_df = bank_df.dropna(subset=["date_idx"]).set_index("date_idx").sort_index()
    bank_df = bank_df.asfreq("Q")
    bank_df = bank_df[bank_df.index <= "2026-03-31"]

    keep_cols = ["Bank", "PCL_t+1", "Year", "Quarter"] + candidate_features_m1 + features_ts_hist
    keep_cols = [c for c in keep_cols if c in bank_df.columns]

    model1_df_bank = bank_df[keep_cols].copy()

    for c in model1_df_bank.columns:
        if c not in ["Bank", "Quarter"]:
            model1_df_bank[c] = pd.to_numeric(model1_df_bank[c], errors="coerce")

    bank_model_data[bank] = model1_df_bank

print(f"\n✅ Setup complete. Built Model 1 datasets for {len(bank_model_data)} banks.\n")
for bank, df_ in bank_model_data.items():
    if "PCL_t" in df_.columns and pd.to_numeric(df_["PCL_t"], errors="coerce").dropna().shape[0] > 0:
        last_pcl = pd.to_numeric(df_["PCL_t"], errors="coerce").dropna().iloc[-1]
        print(f"{bank}: Latest PCL_t = ${last_pcl:,.0f}M")
    else:
        print(f"{bank}: Latest PCL_t = NaN")

# ==============================================================================
# MODEL BACKTEST PER BANK
# ==============================================================================

print("\n>>> Executing Model 1 Pipeline PER BANK...\n")

all_bank_results = {}
all_bank_scorecards = {}
bank_winner_summary = {}
bank_best_model_objects = {}
all_bank_forecasts_1y = {}

for bank, model1_df in bank_model_data.items():
    print("\n" + "=" * 100)
    print(f"🏦 BANK: {bank}")
    print("=" * 100)

    train_valid_df1 = model1_df.dropna(subset=["PCL_t+1"]).copy()

    numeric_cols = [c for c in train_valid_df1.columns if c not in ["Bank", "Quarter"]]
    train_valid_df1[numeric_cols] = train_valid_df1[numeric_cols].apply(pd.to_numeric, errors="coerce")

    if len(train_valid_df1) < 16:
        print(f"⚠️ Skipping {bank}: not enough history for stable backtesting.")
        continue

    # --------------------------------------------------------------------------
    # 1) FEATURE SELECTION
    # --------------------------------------------------------------------------
    fs_feature_pool = [c for c in candidate_features_m1 if c in train_valid_df1.columns]
    fs_df = train_valid_df1.dropna(subset=fs_feature_pool + ["PCL_t+1"]).copy()

    if len(fs_df) < 10:
        selected_features_m1 = [c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1"] if c in train_valid_df1.columns]
    else:
        n_splits_fs = min(3, len(fs_df) - 1)
        if n_splits_fs < 2:
            selected_features_m1 = [c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1"] if c in train_valid_df1.columns]
        else:
            tscv_fs = TimeSeriesSplit(n_splits=n_splits_fs)
            X_fs = fs_df[fs_feature_pool].apply(pd.to_numeric, errors="coerce").astype(float)
            y_full1 = pd.to_numeric(fs_df["PCL_t+1"], errors="coerce").astype(float).values
            X_cand_sc1 = StandardScaler().fit_transform(X_fs)

            elastic_fs1 = ElasticNetCV(
                l1_ratio=[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1.0],
                cv=tscv_fs,
                random_state=42
            )
            elastic_fs1.fit(X_cand_sc1, y_full1)

            selected_features_m1 = []
            print(f"\n🔍 FEATURE SELECTION: {bank}")
            print("-" * 85)
            for feat, coef in zip(fs_feature_pool, elastic_fs1.coef_):
                status = "✅ KEPT" if abs(coef) > 1e-4 else "❌ PRUNED"
                if abs(coef) > 1e-4:
                    selected_features_m1.append(feat)
                print(f"  {status:10s} : {feat:35s} (Weight: {coef:>8.2f})")

            if len(selected_features_m1) <= 1:
                selected_features_m1 = [c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1"] if c in train_valid_df1.columns]

    exog_m1 = [f for f in selected_features_m1 if f != "PCL_t"]
    if not exog_m1:
        exog_m1 = [c for c in ["Fcst_GDP_YoY_for_t+1"] if c in train_valid_df1.columns]

    # --------------------------------------------------------------------------
    # 2) MODEL SETUP
    # --------------------------------------------------------------------------
    n_splits_bt = min(3, len(train_valid_df1) - 1)
    if n_splits_bt < 2:
        print(f"⚠️ Skipping {bank}: insufficient rows for TimeSeriesSplit.")
        continue

    tscv = TimeSeriesSplit(n_splits=n_splits_bt)
    ml_models_dict = make_ml_models_dict(tscv)

    h_test = min(8, max(1, len(train_valid_df1) // 3))
    test_indices = list(range(len(train_valid_df1) - h_test, len(train_valid_df1)))
    prediction_records1 = []

    # --------------------------------------------------------------------------
    # 3) WALK-FORWARD BACKTEST
    # --------------------------------------------------------------------------
    for t in test_indices:
        if t < 12:
            continue

        train_df = train_valid_df1.iloc[:t].copy()
        test_row = train_valid_df1.iloc[t:t+1].copy()

        y_tr_abs = pd.to_numeric(train_df["PCL_t+1"], errors="coerce").dropna()
        y_actual = float(pd.to_numeric(test_row["PCL_t+1"], errors="coerce").iloc[0])

        forecast_date = test_row.index[0]
        yr = int(forecast_date.year)
        qtr = f"Q{forecast_date.quarter}"

        def eval_mod(m_name, p_val):
            prediction_records1.append({
                "Bank": bank,
                "Year": yr,
                "Quarter": qtr,
                "Model": m_name,
                "Predicted ($M)": float(np.squeeze(p_val)) if pd.notna(p_val) else np.nan,
                "Actual ($M)": y_actual
            })

        try:
            baseline_pred = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1]
            eval_mod("01. Baseline: Naive (Last Qtr)", baseline_pred)
        except:
            eval_mod("01. Baseline: Naive (Last Qtr)", np.nan)

        try:
            if HAS_PMDARIMA:
                sarima_model = auto_arima(y_tr_abs, seasonal=False, suppress_warnings=True, error_action="ignore")
                sar_p = sarima_model.predict(n_periods=1)[0]
            else:
                arima_model = ARIMA(y_tr_abs.values, order=(1, 0, 0)).fit()
                sar_p = arima_model.forecast(steps=1)[0]
            eval_mod("02. TS: SARIMA (Blind)", sar_p)
        except:
            fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("02. TS: SARIMA (Blind)", fallback)

        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            var_input = train_df[var_cols].apply(pd.to_numeric, errors="coerce").dropna()
            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_fit = VAR(var_input).fit(maxlags=1)
                var_p = var_fit.forecast(var_input.values[-1:], steps=1)[0, 0]
                eval_mod("03. TS: VAR (Blind)", var_p)
            else:
                fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("03. TS: VAR (Blind)", fallback)
        except:
            fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("03. TS: VAR (Blind)", fallback)

        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            arimax_train = train_df[needed_cols].apply(pd.to_numeric, errors="coerce").dropna()
            arimax_test = test_row[exog_m1].apply(pd.to_numeric, errors="coerce").copy()

            if len(arimax_train) > 5 and not arimax_test.isna().any(axis=None):
                arimax_mod = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=arimax_train[exog_m1].values,
                    order=(1, 0, 0)
                ).fit()
                arimax_fcst = arimax_mod.forecast(steps=1, exog=arimax_test[exog_m1].values)[0]
                eval_mod("04. TS: ARIMAX (Macro-Aware)", arimax_fcst)
            else:
                fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)
        except:
            fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)

        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            vmax_endog = train_df[vmax_endog_cols].apply(pd.to_numeric, errors="coerce").dropna()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_df.loc[vmax_endog.index, exog_m1].apply(pd.to_numeric, errors="coerce")
                valid_idx = vmax_exog.dropna().index.intersection(vmax_endog.index)
                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]
                test_exog = test_row[exog_m1].apply(pd.to_numeric, errors="coerce")

                if len(vmax_endog) > 5 and not test_exog.isna().any(axis=None):
                    vmax_mod = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)
                    vmax_fcst = vmax_mod.forecast(steps=1, exog=test_exog).iloc[0, 0]
                    eval_mod("05. TS: VARMAX (Macro-Aware)", vmax_fcst)
                else:
                    fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
                    eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
            else:
                fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
        except:
            fallback = pd.to_numeric(train_df["PCL_t"], errors="coerce").iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)

        ml_cols = selected_features_m1 + ["PCL_t+1"]
        ml_train = train_df[ml_cols].apply(pd.to_numeric, errors="coerce").dropna().copy() if len(selected_features_m1) > 0 else pd.DataFrame()
        ml_test = test_row[selected_features_m1].apply(pd.to_numeric, errors="coerce").copy() if len(selected_features_m1) > 0 else pd.DataFrame()

        if len(ml_train) > 5 and not ml_test.isna().any(axis=None):
            sc = StandardScaler()
            X_tr_sc = sc.fit_transform(ml_train[selected_features_m1].astype(float))
            y_tr_ml = ml_train["PCL_t+1"].astype(float).values
            X_te_sc = sc.transform(ml_test[selected_features_m1].astype(float))

            for m_name, model in ml_models_dict.items():
                try:
                    fitted_model = model.fit(X_tr_sc, y_tr_ml)
                    pred = fitted_model.predict(X_te_sc)[0]
                    eval_mod(m_name, pred)
                except:
                    eval_mod(m_name, np.nan)
        else:
            for m_name in ml_models_dict.keys():
                eval_mod(m_name, np.nan)

    # --------------------------------------------------------------------------
    # 4) EVALUATION
    # --------------------------------------------------------------------------
    results_df1 = pd.DataFrame(prediction_records1)

    if results_df1.empty:
        print(f"⚠️ No valid OOS predictions for {bank}.")
        continue

    metrics_list1 = []
    for model_name, grp in results_df1.groupby("Model"):
        valid = grp.dropna(subset=["Predicted ($M)", "Actual ($M)"]).copy()
        if len(valid) > 0:
            y_a = valid["Actual ($M)"].values.astype(float)
            y_p = valid["Predicted ($M)"].values.astype(float)
            oos_r2 = r2_score(y_a, y_p) if len(y_a) > 1 else np.nan

            metrics_list1.append({
                "Bank": bank,
                "Model": model_name,
                "OOS R2": float(oos_r2) if pd.notna(oos_r2) else np.nan,
                "OOS MAPE": float(safe_mape(y_a, y_p)),
                "OOS RMSE": float(np.sqrt(np.mean((y_a - y_p) ** 2))),
                "N_OOS": len(valid)
            })

    final_table1 = pd.DataFrame(metrics_list1).sort_values(by="OOS RMSE").reset_index(drop=True)

    all_bank_results[bank] = results_df1
    all_bank_scorecards[bank] = final_table1

    print(f"\n📊 SCORECARD: {bank}")
    display(final_table1)

    # Overall winner stays RMSE-based
    best_row = final_table1.iloc[0]
    best_model_name = best_row["Model"]

    # Best ML model for ML-specific importance uses LOWEST RMSE
    ml_only_rmse = final_table1[final_table1["Model"].str.contains("ML:", na=False)].copy()
    ml_only_rmse = ml_only_rmse.sort_values(by="OOS RMSE").reset_index(drop=True)
    best_ml_rmse_m1 = ml_only_rmse.iloc[0]["Model"] if not ml_only_rmse.empty else None

    bank_winner_summary[bank] = {
        "Bank": bank,
        "Winning Model": best_row["Model"],
        "Best OOS MAPE": best_row["OOS MAPE"],
        "Best OOS RMSE": best_row["OOS RMSE"],
        "Best OOS R2": best_row["OOS R2"],
        "OOS Points": best_row["N_OOS"],
        "Best ML Model by RMSE": best_ml_rmse_m1,
        "Selected Features": ", ".join(selected_features_m1)
    }

    print(f"🏆 Best overall model for {bank}: {best_row['Model']} (selected by lowest OOS MAPE)")
    if best_ml_rmse_m1 is not None:
        print(f"🚀 Best ML model by RMSE for {bank}: {best_ml_rmse_m1}")

    # --------------------------------------------------------------------------
    # 5) FEATURE IMPORTANCE
    #    FIX:
    #    - VAR winner -> TRUE VAR feature importance from fitted VAR coefficients
    #    - ARIMAX / VARMAX winners -> model-based coefficient importance
    #    - ML winners -> ML importance method
    # --------------------------------------------------------------------------
    print(f"\n🔍 Feature Importance for {bank}")
    print("-" * 80)

    ml_cols = selected_features_m1 + ["PCL_t+1"]
    full_ml_df = train_valid_df1[ml_cols].copy()

    for c in full_ml_df.columns:
        full_ml_df[c] = full_ml_df[c].apply(clean_num)
        full_ml_df[c] = pd.to_numeric(full_ml_df[c], errors="coerce")

    full_ml_df = full_ml_df.replace([np.inf, -np.inf], np.nan).dropna().copy()

    if len(train_valid_df1) > 5:

        # ===== CASE A: VAR WINNER =====
        if str(best_model_name) == "03. TS: VAR (Blind)":
            print(f"Using TRUE VAR feature importance for winner: {best_model_name}")

            ts_imp_df = var_feature_importance(
                train_valid_df1=train_valid_df1,
                features_ts_hist=features_ts_hist,
                target_col="PCL_t+1",
                maxlags=1,
                include_self_lag=False
            )

            if not ts_imp_df.empty:
                display(ts_imp_df)
                plot_feature_bar(
                    ts_imp_df,
                    bank,
                    best_model_name,
                    value_col="Feature Importance",
                    xlabel="Absolute VAR Coefficient Sum"
                )
            else:
                print(f"⚠️ Not enough valid data for VAR feature importance for {bank}.")

        # ===== CASE B: ARIMAX WINNER =====
        elif str(best_model_name) == "04. TS: ARIMAX (Macro-Aware)":
            print(f"Using ARIMAX coefficient-based importance for winner: {best_model_name}")

            arimax_imp_df = arimax_feature_importance(
                train_valid_df1=train_valid_df1,
                exog_m1=exog_m1,
                target_col="PCL_t+1",
                order=(1, 0, 0)
            )

            if not arimax_imp_df.empty:
                display(arimax_imp_df)
                plot_feature_bar(
                    arimax_imp_df,
                    bank,
                    best_model_name,
                    value_col="Feature Importance",
                    xlabel="Absolute ARIMAX Coefficient"
                )
            else:
                print(f"⚠️ Not enough valid data for ARIMAX feature importance for {bank}.")

        # ===== CASE C: VARMAX WINNER =====
        elif str(best_model_name) == "05. TS: VARMAX (Macro-Aware)":
            print(f"Using VARMAX coefficient-based importance for winner: {best_model_name}")

            varmax_imp_df = varmax_feature_importance(
                train_valid_df1=train_valid_df1,
                features_ts_hist=features_ts_hist,
                exog_m1=exog_m1,
                target_col="PCL_t+1"
            )

            if not varmax_imp_df.empty:
                display(varmax_imp_df)
                plot_feature_bar(
                    varmax_imp_df,
                    bank,
                    best_model_name,
                    value_col="Feature Importance",
                    xlabel="Absolute VARMAX Exog Coefficient Sum"
                )
            else:
                print(f"⚠️ Not enough valid data for VARMAX feature importance for {bank}.")

        # ===== CASE D: SARIMA / BASELINE =====
        elif str(best_model_name) in ["01. Baseline: Naive (Last Qtr)", "02. TS: SARIMA (Blind)"]:
            print(f"⚠️ {best_model_name} does not have multi-feature model-based importance.")

        # ===== CASE E: BEST MODEL IS ML =====
        elif "ML:" in str(best_model_name):
            if len(full_ml_df) > 5 and len(selected_features_m1) > 0:
                X = full_ml_df[selected_features_m1].astype(float)
                y = full_ml_df["PCL_t+1"].astype(float).values

                sc = StandardScaler()
                X_scaled = sc.fit_transform(X)

                model = ml_models_dict.get(best_model_name)

                if model is not None:
                    try:
                        fitted_model = model.fit(X_scaled, y)
                        bank_best_model_objects[bank] = fitted_model

                        # XGBoost -> built-in importance
                        if "XGBoost" in best_model_name:
                            importance_df = pd.DataFrame({
                                "Feature": selected_features_m1,
                                "Feature Importance": fitted_model.feature_importances_,
                                "Method": "Built-in Gain/Importance"
                            }).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

                            display(importance_df)
                            plot_feature_bar(
                                importance_df,
                                bank,
                                best_model_name,
                                value_col="Feature Importance",
                                xlabel="XGBoost Feature Importance"
                            )

                        # Tree models -> SHAP if available, else permutation importance
                        elif any(k in best_model_name for k in ["Random Forest", "Gradient Boosting", "LightGBM"]):
                            if HAS_SHAP:
                                X_shap = pd.DataFrame(
                                    np.asarray(X_scaled, dtype=np.float64),
                                    columns=selected_features_m1,
                                    index=X.index
                                )

                                explainer = shap.TreeExplainer(fitted_model)
                                shap_values = explainer.shap_values(X_shap)

                                if isinstance(shap_values, list):
                                    shap_values = shap_values[0]

                                if hasattr(shap_values, "ndim") and shap_values.ndim == 3:
                                    shap_values = shap_values[:, :, 0]

                                shap_importance = np.abs(np.asarray(shap_values, dtype=np.float64)).mean(axis=0)

                                shap_df = pd.DataFrame({
                                    "Feature": selected_features_m1,
                                    "Feature Importance": shap_importance,
                                    "Method": "Mean |SHAP|"
                                }).sort_values(by="Feature Importance", ascending=False).reset_index(drop=True)

                                display(shap_df)
                                plot_feature_bar(
                                    shap_df,
                                    bank,
                                    best_model_name,
                                    value_col="Feature Importance",
                                    xlabel="Mean |SHAP value|"
                                )
                            else:
                                perm = permutation_importance(
                                    fitted_model,
                                    X_scaled,
                                    y,
                                    n_repeats=20,
                                    random_state=42
                                )
                                perm_df = pd.DataFrame({
                                    "Feature": selected_features_m1,
                                    "Feature Importance": perm.importances_mean,
                                    "Method": "Permutation Importance"
                                }).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

                                display(perm_df)
                                plot_feature_bar(
                                    perm_df,
                                    bank,
                                    best_model_name,
                                    value_col="Feature Importance",
                                    xlabel="Permutation Importance"
                                )

                        # Linear models -> coefficients
                        elif hasattr(fitted_model, "coef_"):
                            coef_vals = np.ravel(fitted_model.coef_)
                            coef_df = pd.DataFrame({
                                "Feature": selected_features_m1,
                                "Feature Importance": np.abs(coef_vals),
                                "Coefficient": coef_vals,
                                "Method": "Absolute Coefficient"
                            }).sort_values("Feature Importance", ascending=False).reset_index(drop=True)

                            display(coef_df)
                            plot_feature_bar(
                                coef_df[["Feature", "Feature Importance"]],
                                bank,
                                best_model_name,
                                value_col="Feature Importance",
                                xlabel="Absolute Coefficient"
                            )

                        else:
                            print(f"⚠️ No supported importance method found for {best_model_name}.")

                    except Exception as e:
                        print(f"⚠️ Feature importance failed for {bank}: {e}")
                else:
                    print(f"⚠️ Could not find model object for {best_model_name}.")
            else:
                print(f"⚠️ Not enough valid data for ML feature importance for {bank}.")

        else:
            print(f"⚠️ Best model type not recognized for feature importance: {best_model_name}")

    else:
        print(f"⚠️ Not enough valid data for feature importance for {bank}.")

    # --------------------------------------------------------------------------
    # 6) NEXT 1 YEAR FORECAST FOR BEST OVERALL MODEL (MAPE-BASED)
    # --------------------------------------------------------------------------
    print(f"\n📈 NEXT 1 YEAR FORECAST FOR {bank} USING BEST MODEL: {best_model_name}")
    print("-" * 100)

    forecast_1y_df = forecast_best_model_1y(
        bank=bank,
        best_model_name=best_model_name,
        train_valid_df1=train_valid_df1,
        bank_df_full=model1_df,
        selected_features_m1=selected_features_m1,
        exog_m1=exog_m1,
        features_ts_hist=features_ts_hist,
        horizon=4
    )

    all_bank_forecasts_1y[bank] = forecast_1y_df
    display(forecast_1y_df)

# ==============================================================================
# 7) COMBINED WINNER SUMMARY
# ==============================================================================

if len(bank_winner_summary) > 0:
    summary_df = pd.DataFrame(bank_winner_summary.values()).sort_values(by="Best OOS RMSE").reset_index(drop=True)

    print("\n" + "=" * 100)
    print("FINAL BANK WINNER SUMMARY")
    print("=" * 100)
    display(summary_df)
else:
    print("⚠️ No bank-level summaries were produced.")

# ==============================================================================
# 8) COMBINED NEXT 1 YEAR FORECAST SUMMARY
# ==============================================================================

if len(all_bank_forecasts_1y) > 0:
    combined_1y_forecast_df = pd.concat(all_bank_forecasts_1y.values(), axis=0).reset_index(drop=True)

    print("\n" + "=" * 100)
    print("COMBINED NEXT 1 YEAR FORECAST SUMMARY (BEST MODEL PER BANK)")
    print("=" * 100)
    display(combined_1y_forecast_df)

    combined_1y_forecast_pivot = combined_1y_forecast_df.pivot_table(
        index=["Bank", "Best Model"],
        columns="Forecast Quarter",
        values="Forecast ($M)"
    ).reset_index()

    print("\n" + "=" * 100)
    print("PIVOT TABLE: NEXT 1 YEAR FORECAST (BEST MODEL PER BANK)")
    print("=" * 100)
    display(combined_1y_forecast_pivot)
else:
    print("⚠️ No next-1-year forecasts were produced.")
```

## UC2 Best model(2021Q4-2025Q4) for BMO, CIBC ##

```python
# ==============================================================================
# MODEL 1 (EARNINGS DAY) - PER-BANK FULL PIPELINE
# BANKS USED: BMO, CIBC ONLY
#
# Updated:
# - runs only for BMO and CIBC
# - Sent_BART_t removed everywhere
# - keep data from 2021Q4 onward only
# - fixed OOS horizon to 4
# - standard scaling applied using TRAIN WINDOW ONLY
# - feature selection moved INSIDE walk-forward loop to avoid OOS leakage
# - duplicate-safe feature handling added
# - SHAP feature importance bar chart added for best ML model (lowest OOS RMSE) per bank
# - next 1 year forecast (next 4 quarters) for the best model (lowest OOS RMSE)
# - forecast tables per bank + combined forecast table
# - NO LOG TRANSFORM on PCL_t or PCL_t+1
# ==============================================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.linear_model import ElasticNetCV, RidgeCV, LassoCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import TimeSeriesSplit

# Optional packages
try:
    from pmdarima import auto_arima
    HAS_PMDARIMA = True
except ImportError:
    HAS_PMDARIMA = False

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

try:
    from lightgbm import LGBMRegressor
    HAS_LGBM = True
except ImportError:
    HAS_LGBM = False

try:
    import shap
    HAS_SHAP = True
except ImportError:
    HAS_SHAP = False

PD_VER = int(pd.__version__.split(".")[0])
Q_FREQ = "QE" if PD_VER >= 2 else "Q"

print(">>> Building feature pool for Model 1 (Per-Bank Earnings Day Forecasts)...")
if not HAS_XGB or not HAS_LGBM:
    print("⚠️ Notice: xgboost or lightgbm not found. Run '!pip install xgboost lightgbm' if needed.")
if not HAS_SHAP:
    print("⚠️ Notice: shap not found. Run '!pip install shap' if needed.")

# ==============================================================================
# HELPERS
# ==============================================================================

def clean_num(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    s = str(x).strip().replace(",", "").replace("$", "").replace(" ", "")
    if s in ["—", "", "-", "NA", "N/A", "nan", "NaN", "Unnamed:23", "Unnamed:39"]:
        return np.nan
    if s.startswith("(") and s.endswith(")"):
        s = "-" + s[1:-1]
    try:
        return float(s)
    except:
        return np.nan

def safe_mape(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    denom = np.where(np.abs(y_true) < 1e-8, np.nan, np.abs(y_true))
    return np.nanmean(np.abs((y_true - y_pred) / denom)) * 100

def unique_keep_order(seq):
    return list(dict.fromkeys(seq))

def scale_train_test(train_df, test_df, cols):
    cols = unique_keep_order(cols)

    scaler = StandardScaler()

    train_vals = scaler.fit_transform(train_df[cols])
    test_vals = scaler.transform(test_df[cols])

    train_scaled = pd.DataFrame(
        train_vals,
        index=train_df.index,
        columns=cols
    )
    test_scaled = pd.DataFrame(
        test_vals,
        index=test_df.index,
        columns=cols
    )
    return train_scaled, test_scaled, scaler

def select_features_train_only(train_df, candidate_features, target_col="PCL_t+1"):
    candidate_features = unique_keep_order(candidate_features)
    fs_feature_pool = [c for c in candidate_features if c in train_df.columns]
    fs_df = train_df.dropna(subset=fs_feature_pool + [target_col]).copy()

    fallback = [c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1"] if c in train_df.columns]
    fallback = unique_keep_order(fallback)

    if len(fs_feature_pool) == 0:
        return fallback

    if len(fs_df) < 6:
        return fallback

    n_splits_fs = min(3, len(fs_df) - 1)
    if n_splits_fs < 2:
        return fallback

    try:
        tscv_fs = TimeSeriesSplit(n_splits=n_splits_fs)

        scaler_fs = StandardScaler()
        X_fs = scaler_fs.fit_transform(fs_df[fs_feature_pool])
        y_fs = fs_df[target_col].values

        elastic_fs = ElasticNetCV(
            l1_ratio=[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1.0],
            cv=tscv_fs,
            random_state=42
        )
        elastic_fs.fit(X_fs, y_fs)

        selected = []
        for feat, coef in zip(fs_feature_pool, elastic_fs.coef_):
            if abs(coef) > 1e-4:
                selected.append(feat)

        selected = unique_keep_order(selected)

        if len(selected) <= 1:
            return fallback

        return selected

    except:
        return fallback

def build_ml_models_dict(tscv):
    models = {
        "06. ML: Lasso": LassoCV(alphas=np.logspace(-2, 4, 50), cv=tscv),
        "07. ML: Ridge": RidgeCV(alphas=np.logspace(-2, 4, 50)),
        "08. ML: Random Forest": RandomForestRegressor(
            n_estimators=100,
            max_depth=3,
            random_state=42
        ),
        "09. ML: Gradient Boosting": GradientBoostingRegressor(
            n_estimators=100,
            random_state=42
        ),
        "10. ML: Elastic Net": ElasticNetCV(
            l1_ratio=[0.1, 0.5, 0.9],
            cv=tscv,
            random_state=42
        )
    }

    if HAS_XGB:
        models["11. ML: XGBoost"] = XGBRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42
        )

    if HAS_LGBM:
        models["12. ML: LightGBM"] = LGBMRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42,
            verbose=-1
        )

    return models

def compute_shap_importance(fitted_model, X_df, model_name, bank_name, max_background=100):
    if not HAS_SHAP:
        print(f"⚠️ SHAP not installed. Skipping SHAP importance for {bank_name}.")
        return None

    if X_df is None or X_df.empty:
        print(f"⚠️ No feature matrix available for SHAP for {bank_name}.")
        return None

    try:
        X_used = X_df.copy()

        if len(X_used) > max_background:
            X_background = X_used.sample(max_background, random_state=42)
        else:
            X_background = X_used.copy()

        if any(k in model_name for k in ["Random Forest", "Gradient Boosting", "XGBoost", "LightGBM"]):
            explainer = shap.TreeExplainer(fitted_model)
            shap_values = explainer.shap_values(X_used)
        elif any(k in model_name for k in ["Lasso", "Ridge", "Elastic Net"]):
            try:
                explainer = shap.LinearExplainer(fitted_model, X_background)
                shap_values = explainer.shap_values(X_used)
            except:
                explainer = shap.Explainer(fitted_model, X_background)
                shap_values = explainer(X_used).values
        else:
            explainer = shap.KernelExplainer(fitted_model.predict, X_background)
            shap_values = explainer.shap_values(X_used, nsamples=100)

        if isinstance(shap_values, list):
            shap_values = shap_values[0]

        if hasattr(shap_values, "values"):
            shap_values = shap_values.values

        shap_values = np.array(shap_values)

        if shap_values.ndim == 1:
            shap_values = shap_values.reshape(-1, 1)

        mean_abs_shap = np.abs(shap_values).mean(axis=0)

        shap_importance_df = pd.DataFrame({
            "Bank": bank_name,
            "Feature": X_used.columns,
            "MeanAbsSHAP": mean_abs_shap
        }).sort_values("MeanAbsSHAP", ascending=False).reset_index(drop=True)

        print("\n" + "-" * 80)
        print(f"SHAP FEATURE IMPORTANCE | {bank_name} | {model_name}")
        print("-" * 80)
        display(shap_importance_df)

        plt.figure(figsize=(9, 5))
        plt.barh(
            shap_importance_df["Feature"][::-1],
            shap_importance_df["MeanAbsSHAP"][::-1]
        )
        plt.title(f"SHAP Feature Importance - {bank_name}\n{model_name}")
        plt.xlabel("Mean |SHAP value|")
        plt.ylabel("Feature")
        plt.tight_layout()
        plt.show()

        return shap_importance_df

    except Exception as e:
        print(f"⚠️ SHAP failed for {bank_name} ({model_name}): {e}")
        return None

def make_future_quarter_index(last_date, steps=4):
    return pd.date_range(
        start=last_date + pd.offsets.QuarterEnd(1),
        periods=steps,
        freq="Q"
    )

def quarter_label(dt):
    return f"{dt.year}Q{dt.quarter}"

def build_future_feature_frame(hist_df, future_index, feature_cols):
    feature_cols = [c for c in unique_keep_order(feature_cols) if c in hist_df.columns]
    out = pd.DataFrame(index=future_index, columns=feature_cols, dtype=float)

    for col in feature_cols:
        for dt in future_index:
            if dt in hist_df.index and pd.notna(hist_df.loc[dt, col]):
                out.loc[dt, col] = hist_df.loc[dt, col]
            else:
                s = hist_df[col].dropna()
                out.loc[dt, col] = s.iloc[-1] if len(s) > 0 else np.nan

    out = out.ffill().bfill()
    return out

def forecast_best_model_next_1y(
    bank,
    best_model_name,
    train_valid_df1,
    model1_df,
    selected_features_final,
    features_ts_hist,
    horizon=4
):
    last_train_date = train_valid_df1.index.max()
    future_index = make_future_quarter_index(last_train_date, steps=horizon)
    future_labels = [quarter_label(d) for d in future_index]

    y_full = train_valid_df1["PCL_t+1"].dropna().copy()
    last_level = float(y_full.iloc[-1]) if len(y_full) > 0 else np.nan

    exog_m1 = [f for f in selected_features_final if f != "PCL_t"]
    exog_m1 = unique_keep_order(exog_m1)

    def make_output(pred_levels):
        return pd.DataFrame({
            "Bank": bank,
            "Forecast Quarter": future_labels,
            "Best Model": best_model_name,
            "Forecast ($M)": pred_levels
        })

    # 01. Baseline
    if best_model_name == "01. Baseline: Naive (Last Qtr)":
        return make_output([last_level] * horizon)

    # 02. SARIMA / ARIMA
    elif best_model_name == "02. TS: SARIMA (Blind)":
        try:
            if HAS_PMDARIMA:
                sarima_model = auto_arima(
                    y_full,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
                preds = sarima_model.predict(n_periods=horizon)
            else:
                sarima_model = ARIMA(y_full.values, order=(1, 0, 0)).fit()
                preds = sarima_model.forecast(steps=horizon)

            preds_level = [float(x) for x in preds]
            return make_output(preds_level)
        except:
            return make_output([last_level] * horizon)

    # 03. VAR
    elif best_model_name == "03. TS: VAR (Blind)":
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            var_cols = unique_keep_order(var_cols)
            var_input = train_valid_df1[var_cols].dropna().copy()

            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_train_scaled, _, _ = scale_train_test(var_input, var_input.tail(1), var_cols)
                var_fit = VAR(var_train_scaled).fit(maxlags=1)

                current_values = var_train_scaled.values.copy()
                preds = []

                y_mean = var_input["PCL_t+1"].mean()
                y_std = var_input["PCL_t+1"].std(ddof=0)

                last_macro_scaled = current_values[-1, 1:] if current_values.shape[1] > 1 else np.array([])

                for _ in range(horizon):
                    fc_scaled = var_fit.forecast(current_values[-var_fit.k_ar:], steps=1)[0]

                    if len(last_macro_scaled) > 0:
                        fc_scaled[1:] = last_macro_scaled

                    if pd.isna(y_std) or abs(y_std) < 1e-12:
                        pred = y_mean
                    else:
                        pred = fc_scaled[0] * y_std + y_mean

                    preds.append(pred)
                    current_values = np.vstack([current_values, fc_scaled])

                preds_level = [float(x) for x in preds]
                return make_output(preds_level)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 04. ARIMAX
    elif best_model_name == "04. TS: ARIMAX (Macro-Aware)":
        try:
            if len(exog_m1) == 0:
                return make_output([last_level] * horizon)

            needed_cols = ["PCL_t+1"] + exog_m1
            needed_cols = unique_keep_order(needed_cols)
            arimax_train = train_valid_df1[needed_cols].dropna().copy()

            if len(arimax_train) > 5:
                future_exog = build_future_feature_frame(model1_df, future_index, exog_m1)

                exog_train_scaled, exog_future_scaled, _ = scale_train_test(
                    arimax_train,
                    future_exog,
                    exog_m1
                )

                arimax_mod = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=exog_train_scaled.values,
                    order=(1, 0, 0)
                ).fit()

                preds = arimax_mod.forecast(steps=horizon, exog=exog_future_scaled.values)
                preds_level = [float(x) for x in preds]
                return make_output(preds_level)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 05. VARMAX
    elif best_model_name == "05. TS: VARMAX (Macro-Aware)":
        try:
            if len(exog_m1) == 0:
                return make_output([last_level] * horizon)

            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            vmax_endog_cols = unique_keep_order(vmax_endog_cols)

            vmax_endog = train_valid_df1[vmax_endog_cols].dropna().copy()
            vmax_exog = train_valid_df1.loc[vmax_endog.index, exog_m1].copy()

            valid_idx = vmax_endog.index.intersection(vmax_exog.dropna().index)
            vmax_endog = vmax_endog.loc[valid_idx]
            vmax_exog = vmax_exog.loc[valid_idx]

            if len(vmax_endog) > 5:
                future_exog = build_future_feature_frame(model1_df, future_index, exog_m1)

                endog_train_scaled, _, _ = scale_train_test(vmax_endog, vmax_endog.tail(1), vmax_endog_cols)
                exog_train_scaled, exog_future_scaled, _ = scale_train_test(vmax_exog, future_exog, exog_m1)

                vmax_mod = VARMAX(
                    endog_train_scaled,
                    exog=exog_train_scaled,
                    order=(1, 0)
                ).fit(disp=False, maxiter=200)

                fc_scaled = vmax_mod.forecast(steps=horizon, exog=exog_future_scaled)
                y_mean = vmax_endog["PCL_t+1"].mean()
                y_std = vmax_endog["PCL_t+1"].std(ddof=0)

                if pd.isna(y_std) or abs(y_std) < 1e-12:
                    preds = [y_mean] * horizon
                else:
                    preds = [fc_scaled.iloc[i, 0] * y_std + y_mean for i in range(horizon)]

                preds_level = [float(x) for x in preds]
                return make_output(preds_level)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 06+ ML
    elif "ML:" in best_model_name:
        try:
            if len(selected_features_final) == 0:
                return make_output([last_level] * horizon)

            ml_final_cols = unique_keep_order(selected_features_final + ["PCL_t+1"])
            ml_final_df = train_valid_df1[ml_final_cols].dropna().copy()

            if len(ml_final_df) < 6:
                return make_output([last_level] * horizon)

            n_splits_final = min(3, len(ml_final_df) - 1)
            if n_splits_final < 2:
                return make_output([last_level] * horizon)

            tscv_final = TimeSeriesSplit(n_splits=n_splits_final)
            ml_models_final = build_ml_models_dict(tscv_final)

            if best_model_name not in ml_models_final:
                return make_output([last_level] * horizon)

            X_train = ml_final_df[selected_features_final].copy()
            y_train = ml_final_df["PCL_t+1"].values

            X_train_scaled, _, scaler_final = scale_train_test(X_train, X_train.tail(1), selected_features_final)

            final_model = ml_models_final[best_model_name]
            final_model.fit(X_train_scaled.values, y_train)

            future_features = build_future_feature_frame(model1_df, future_index, selected_features_final).copy()

            preds = []
            last_pred = y_train[-1]

            for i in range(horizon):
                row = future_features.iloc[i:i+1].copy()

                if "PCL_t" in row.columns:
                    row.loc[:, "PCL_t"] = last_pred

                row_scaled = pd.DataFrame(
                    scaler_final.transform(row[selected_features_final]),
                    columns=selected_features_final,
                    index=row.index
                )

                pred = final_model.predict(row_scaled.values)[0]
                preds.append(float(pred))
                last_pred = pred

            preds_level = [float(x) for x in preds]
            return make_output(preds_level)

        except Exception as e:
            print(f"⚠️ Forecasting failed for {bank} | {best_model_name}: {e}")
            return make_output([last_level] * horizon)

    else:
        return make_output([last_level] * horizon)

# ==============================================================================
# LOAD DATA
# ==============================================================================

file_hist = "Complete_UC2_Forecast_Data.xlsx"

raw_df = pd.read_excel(file_hist, sheet_name="Forecast Data", header=0)
raw_df.columns = [str(c).strip() for c in raw_df.columns]

print("\nDetected columns:")
print(raw_df.columns.tolist())

required_base_cols = ["Bank", "Year", "Quarter"]
missing_base_cols = [c for c in required_base_cols if c not in raw_df.columns]
if missing_base_cols:
    raise ValueError(f"Missing required columns: {missing_base_cols}")

cols_to_clean = [
    "PCL_t+1",
    "PCL_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1",
    "GDP_YoY_t",
    "Overnight_Rate_t"
]

for c in cols_to_clean:
    if c in raw_df.columns:
        raw_df[c] = raw_df[c].apply(clean_num)

raw_df["Bank"] = raw_df["Bank"].astype(str).str.strip()
raw_df["Year"] = pd.to_numeric(raw_df["Year"], errors="coerce")
raw_df["Quarter"] = raw_df["Quarter"].astype(str).str.strip()

raw_df = raw_df.dropna(subset=["Bank", "Year", "Quarter"])
raw_df = raw_df.sort_values(by=["Bank", "Year", "Quarter"]).reset_index(drop=True)

# keep BMO and CIBC only
target_banks = ["BMO", "CIBC"]
raw_df = raw_df[raw_df["Bank"].isin(target_banks)].copy()

cols_to_fill = [
    "PCL_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

for c in cols_to_fill:
    if c in raw_df.columns:
        raw_df[c] = raw_df.groupby("Bank")[c].transform(lambda x: x.ffill().bfill())

q_map = {"Q1": "03-31", "Q2": "06-30", "Q3": "09-30", "Q4": "12-31"}

features_ts_hist = ["GDP_YoY_t", "Overnight_Rate_t"]

# ==============================================================================
# BUILD PER-BANK DATASETS
# ==============================================================================

bank_model_data = {}

for bank in target_banks:
    bank_df = raw_df[raw_df["Bank"] == bank].copy()

    if bank_df.empty:
        print(f"⚠️ No data found for {bank}. Skipping.")
        continue

    bank_df["date_idx"] = pd.to_datetime(
        bank_df["Year"].astype(int).astype(str) + "-" + bank_df["Quarter"].map(q_map),
        errors="coerce"
    )

    bank_df = bank_df.dropna(subset=["date_idx"]).set_index("date_idx").sort_index()

    # Keep 2021Q4 onward only
    bank_df = bank_df[bank_df.index >= "2021-12-31"]
    bank_df = bank_df[bank_df.index <= "2026-03-31"]

    candidate_features_m1 = [
        "PCL_t",
        "Fcst_GDP_YoY_for_t+1",
        "Fcst_Unemp_A_for_t+1",
        "Fcst_BoC_for_t+1"
    ]
    candidate_features_m1 = unique_keep_order(candidate_features_m1)

    keep_cols = [
        "Bank", "PCL_t+1", "PCL_t", "Year", "Quarter"
    ] + candidate_features_m1 + features_ts_hist
    keep_cols = [c for c in unique_keep_order(keep_cols) if c in bank_df.columns]

    model1_df_bank = bank_df[keep_cols].copy()
    bank_model_data[bank] = model1_df_bank

print(f"\n✅ Setup complete. Built Model 1 datasets for {len(bank_model_data)} banks.\n")
for bank, df_ in bank_model_data.items():
    if "PCL_t" in df_.columns and df_["PCL_t"].dropna().shape[0] > 0:
        last_pcl = df_["PCL_t"].dropna().iloc[-1]
        print(f"{bank}: Latest PCL_t = ${last_pcl:,.0f}M")
    else:
        print(f"{bank}: Latest PCL_t = NaN")

# ==============================================================================
# MODEL BACKTEST PER BANK
# ==============================================================================

print("\n>>> Executing Model 1 Pipeline PER BANK...\n")

all_bank_results = {}
all_bank_scorecards = {}
bank_winner_summary = []
all_bank_shap = {}
all_bank_forecasts = {}
final_selected_features_by_bank = {}

for bank, model1_df in bank_model_data.items():
    print("\n" + "=" * 100)
    print(f"🏦 BANK: {bank}")
    print("=" * 100)

    train_valid_df1 = model1_df.dropna(subset=["PCL_t+1"]).copy()

    if len(train_valid_df1) < 8:
        print(f"⚠️ Skipping {bank}: not enough history after 2021Q4 filter.")
        continue

    h_test = 4
    if len(train_valid_df1) <= h_test:
        print(f"⚠️ Skipping {bank}: not enough rows to reserve 4 OOS periods.")
        continue

    ml_model_names = list(build_ml_models_dict(tscv=TimeSeriesSplit(n_splits=2)).keys())

    test_indices = list(range(len(train_valid_df1) - h_test, len(train_valid_df1)))
    prediction_records1 = []

    for t in test_indices:
        if t < 4:
            continue

        train_df = train_valid_df1.iloc[:t].copy()
        test_row = train_valid_df1.iloc[t:t+1].copy()

        y_tr = train_df["PCL_t+1"]
        y_actual_level = float(test_row["PCL_t+1"].iloc[0])

        forecast_date = test_row.index[0]
        yr = int(forecast_date.year)
        qtr = f"Q{forecast_date.quarter}"

        candidate_features_m1 = [
            "PCL_t",
            "Fcst_GDP_YoY_for_t+1",
            "Fcst_Unemp_A_for_t+1",
            "Fcst_BoC_for_t+1"
        ]
        candidate_features_m1 = unique_keep_order(candidate_features_m1)

        selected_features_m1 = select_features_train_only(
            train_df,
            candidate_features_m1,
            target_col="PCL_t+1"
        )
        selected_features_m1 = unique_keep_order(selected_features_m1)

        exog_m1 = [f for f in selected_features_m1 if f != "PCL_t"]
        exog_m1 = unique_keep_order(exog_m1)

        if not exog_m1:
            exog_m1 = [c for c in ["Fcst_GDP_YoY_for_t+1"] if c in train_df.columns]
            exog_m1 = unique_keep_order(exog_m1)

        print(f"{bank} | {yr} {qtr} | Selected features: {selected_features_m1}")

        def eval_mod(m_name, pred_value):
            pred_level = float(np.squeeze(pred_value)) if pd.notna(pred_value) else np.nan
            prediction_records1.append({
                "Bank": bank,
                "Year": yr,
                "Quarter": qtr,
                "Model": m_name,
                "Predicted ($M)": pred_level,
                "Actual ($M)": y_actual_level
            })

        # 01. Baseline
        try:
            baseline_pred = train_df["PCL_t"].iloc[-1]
            eval_mod("01. Baseline: Naive (Last Qtr)", baseline_pred)
        except:
            eval_mod("01. Baseline: Naive (Last Qtr)", np.nan)

        # 02. SARIMA / ARIMA on level target
        try:
            if HAS_PMDARIMA:
                sarima_model = auto_arima(
                    y_tr,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
                sar_p = sarima_model.predict(n_periods=1)[0]
            else:
                sarima_model = ARIMA(y_tr.values, order=(1, 0, 0)).fit()
                sar_p = sarima_model.forecast(steps=1)[0]

            eval_mod("02. TS: SARIMA (Blind)", sar_p)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("02. TS: SARIMA (Blind)", fallback)

        # 03. VAR on level target + scaled train window only
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            var_cols = unique_keep_order(var_cols)
            var_input = train_df[var_cols].dropna()

            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_train_scaled, _, _ = scale_train_test(var_input, var_input.tail(1), var_cols)
                var_fit = VAR(var_train_scaled).fit(maxlags=1)
                var_p_scaled = var_fit.forecast(var_train_scaled.values[-1:], steps=1)[0, 0]

                y_mean = var_input["PCL_t+1"].mean()
                y_std = var_input["PCL_t+1"].std(ddof=0)

                if pd.isna(y_std) or abs(y_std) < 1e-12:
                    var_p = y_mean
                else:
                    var_p = var_p_scaled * y_std + y_mean

                eval_mod("03. TS: VAR (Blind)", var_p)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("03. TS: VAR (Blind)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("03. TS: VAR (Blind)", fallback)

        # 04. ARIMAX on level target, scaled exog train-only
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            needed_cols = unique_keep_order(needed_cols)

            arimax_train = train_df[needed_cols].dropna()
            arimax_test = test_row[exog_m1].copy()

            if len(arimax_train) > 5 and len(exog_m1) > 0 and not arimax_test.isna().any(axis=None):
                exog_train_scaled, exog_test_scaled, _ = scale_train_test(
                    arimax_train,
                    arimax_test,
                    exog_m1
                )

                arimax_mod = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=exog_train_scaled.values,
                    order=(1, 0, 0)
                ).fit()

                arimax_fcst = arimax_mod.forecast(steps=1, exog=exog_test_scaled.values)[0]
                eval_mod("04. TS: ARIMAX (Macro-Aware)", arimax_fcst)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)

        # 05. VARMAX on level target, scaled endog/exog train-only
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            vmax_endog_cols = unique_keep_order(vmax_endog_cols)

            vmax_endog = train_df[vmax_endog_cols].dropna()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_df.loc[vmax_endog.index, exog_m1]
                valid_idx = vmax_exog.dropna().index.intersection(vmax_endog.index)

                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]
                test_exog = test_row[exog_m1]

                if len(vmax_endog) > 5 and not test_exog.isna().any(axis=None):
                    endog_train_scaled, _, _ = scale_train_test(vmax_endog, vmax_endog.tail(1), vmax_endog_cols)
                    exog_train_scaled, exog_test_scaled, _ = scale_train_test(vmax_exog, test_exog, exog_m1)

                    vmax_mod = VARMAX(
                        endog_train_scaled,
                        exog=exog_train_scaled,
                        order=(1, 0)
                    ).fit(disp=False, maxiter=200)

                    vmax_fcst_scaled = vmax_mod.forecast(steps=1, exog=exog_test_scaled).iloc[0, 0]

                    y_mean = vmax_endog["PCL_t+1"].mean()
                    y_std = vmax_endog["PCL_t+1"].std(ddof=0)

                    if pd.isna(y_std) or abs(y_std) < 1e-12:
                        vmax_fcst = y_mean
                    else:
                        vmax_fcst = vmax_fcst_scaled * y_std + y_mean

                    eval_mod("05. TS: VARMAX (Macro-Aware)", vmax_fcst)
                else:
                    fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                    eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)

        # 06+ ML on level target and level PCL_t
        ml_cols = unique_keep_order(selected_features_m1 + ["PCL_t+1"])
        ml_train = train_df[ml_cols].dropna().copy() if len(selected_features_m1) > 0 else pd.DataFrame()
        ml_test = test_row[selected_features_m1].copy() if len(selected_features_m1) > 0 else pd.DataFrame()

        n_splits_bt = min(3, len(ml_train) - 1) if len(ml_train) > 2 else 0

        if len(ml_train) > 5 and not ml_test.isna().any(axis=None) and n_splits_bt >= 2:
            X_train_scaled, X_test_scaled, _ = scale_train_test(ml_train, ml_test, selected_features_m1)
            y_tr_ml = ml_train["PCL_t+1"].values

            tscv = TimeSeriesSplit(n_splits=n_splits_bt)
            ml_models_dict = build_ml_models_dict(tscv)

            for m_name, model in ml_models_dict.items():
                try:
                    fitted_model = model.fit(X_train_scaled.values, y_tr_ml)
                    pred = fitted_model.predict(X_test_scaled.values)[0]
                    eval_mod(m_name, pred)
                except:
                    eval_mod(m_name, np.nan)
        else:
            for m_name in ml_model_names:
                eval_mod(m_name, np.nan)

    # --------------------------------------------------------------------------
    # 4) EVALUATION
    # --------------------------------------------------------------------------
    results_df1 = pd.DataFrame(prediction_records1)

    if results_df1.empty:
        print(f"⚠️ No valid OOS predictions for {bank}.")
        continue

    metrics_list1 = []
    for model_name, grp in results_df1.groupby("Model"):
        valid = grp.dropna(subset=["Predicted ($M)", "Actual ($M)"]).copy()

        if len(valid) > 0:
            y_a = valid["Actual ($M)"].values.astype(float)
            y_p = valid["Predicted ($M)"].values.astype(float)

            oos_r2 = r2_score(y_a, y_p) if len(valid) > 1 else np.nan

            metrics_list1.append({
                "Bank": bank,
                "Model": model_name,
                "OOS R2": oos_r2,
                "OOS MAPE": float(safe_mape(y_a, y_p)),
                "OOS RMSE": float(np.sqrt(np.mean((y_a - y_p) ** 2))),
                "N_OOS": len(valid)
            })

    final_table1 = pd.DataFrame(metrics_list1).sort_values(
        by=["OOS RMSE", "OOS MAPE"],
        ascending=[True, True]
    ).reset_index(drop=True)

    all_bank_results[bank] = results_df1
    all_bank_scorecards[bank] = final_table1

    print("\n" + "-" * 100)
    print(f"SCORECARD | {bank}")
    print("-" * 100)
    display(final_table1)

    best_row = final_table1.iloc[0]
    best_model_name = best_row["Model"]

    ml_only = final_table1[final_table1["Model"].str.contains("ML:", na=False)].copy()
    if not ml_only.empty:
        ml_only = ml_only.sort_values(by=["OOS RMSE", "OOS MAPE"], ascending=[True, True]).reset_index(drop=True)
        best_ml_m1 = ml_only.iloc[0]["Model"]
    else:
        best_ml_m1 = None

    # use full training data to choose final features for forecasting / shap
    candidate_features_m1 = [
        "PCL_t",
        "Fcst_GDP_YoY_for_t+1",
        "Fcst_Unemp_A_for_t+1",
        "Fcst_BoC_for_t+1"
    ]
    candidate_features_m1 = unique_keep_order(candidate_features_m1)

    selected_features_final = select_features_train_only(
        train_valid_df1,
        candidate_features_m1,
        target_col="PCL_t+1"
    )
    selected_features_final = unique_keep_order(selected_features_final)
    final_selected_features_by_bank[bank] = selected_features_final

    bank_winner_summary.append({
        "Bank": bank,
        "Winning Model": best_row["Model"],
        "Best OOS RMSE": best_row["OOS RMSE"],
        "Best OOS MAPE": best_row["OOS MAPE"],
        "Best OOS R2": best_row["OOS R2"],
        "OOS Points": best_row["N_OOS"],
        "Best ML Model": best_ml_m1,
        "Final Selected Features": ", ".join(selected_features_final)
    })

    print(f"🏆 Best overall model for {bank}: {best_row['Model']}")
    if best_ml_m1 is not None:
        print(f"🚀 Best ML model for {bank}: {best_ml_m1}")

    # --------------------------------------------------------------------------
    # 5) SHAP FEATURE IMPORTANCE FOR BEST ML MODEL
    # --------------------------------------------------------------------------
    if best_ml_m1 is None:
        print(f"⚠️ No ML model available for SHAP for {bank}.")
    else:
        try:
            full_train_df = train_valid_df1.copy()

            ml_final_cols = unique_keep_order(selected_features_final + ["PCL_t+1"])
            ml_final_df = full_train_df[ml_final_cols].dropna().copy()

            if len(ml_final_df) < 6 or len(selected_features_final) == 0:
                print(f"⚠️ Not enough clean data for SHAP on {bank}.")
            else:
                X_final = ml_final_df[selected_features_final].copy()
                y_final = ml_final_df["PCL_t+1"].values

                n_splits_final = min(3, len(ml_final_df) - 1)
                if n_splits_final < 2:
                    print(f"⚠️ Not enough rows to fit final SHAP model for {bank}.")
                else:
                    tscv_final = TimeSeriesSplit(n_splits=n_splits_final)
                    ml_models_final = build_ml_models_dict(tscv_final)

                    if best_ml_m1 not in ml_models_final:
                        print(f"⚠️ Best ML model {best_ml_m1} not found for {bank}.")
                    else:
                        X_final_scaled, _, scaler_final = scale_train_test(X_final, X_final.tail(1), selected_features_final)
                        final_model = ml_models_final[best_ml_m1]
                        final_model.fit(X_final_scaled.values, y_final)

                        shap_importance_df = compute_shap_importance(
                            fitted_model=final_model,
                            X_df=X_final_scaled,
                            model_name=best_ml_m1,
                            bank_name=bank
                        )

                        if shap_importance_df is not None:
                            all_bank_shap[bank] = shap_importance_df

        except Exception as e:
            print(f"⚠️ SHAP block failed for {bank}: {e}")

    # --------------------------------------------------------------------------
    # 6) NEXT 1 YEAR FORECAST FOR BEST MODEL
    # --------------------------------------------------------------------------
    try:
        forecast_df = forecast_best_model_next_1y(
            bank=bank,
            best_model_name=best_model_name,
            train_valid_df1=train_valid_df1,
            model1_df=model1_df,
            selected_features_final=selected_features_final,
            features_ts_hist=features_ts_hist,
            horizon=4
        )

        all_bank_forecasts[bank] = forecast_df

        print("\n" + "-" * 100)
        print(f"NEXT 1 YEAR FORECAST | {bank} | BEST MODEL = {best_model_name}")
        print("-" * 100)
        display(forecast_df)

    except Exception as e:
        print(f"⚠️ Forecast block failed for {bank}: {e}")

# ==============================================================================
# 7) COMBINED WINNER SUMMARY
# ==============================================================================

if len(bank_winner_summary) > 0:
    summary_df = pd.DataFrame(bank_winner_summary).sort_values(
        by=["Best OOS RMSE", "Best OOS MAPE"],
        ascending=[True, True]
    ).reset_index(drop=True)

    print("\n" + "=" * 100)
    print("FINAL BANK WINNER SUMMARY")
    print("=" * 100)
    display(summary_df)
else:
    print("⚠️ No bank-level summaries were produced.")

# ==============================================================================
# 8) COMBINED FORECAST TABLE
# ==============================================================================

if len(all_bank_forecasts) > 0:
    combined_forecast_df = pd.concat(all_bank_forecasts.values(), axis=0).reset_index(drop=True)

    print("\n" + "=" * 100)
    print("COMBINED NEXT 1 YEAR FORECAST TABLE | BMO + CIBC")
    print("=" * 100)
    display(combined_forecast_df)

    forecast_pivot_df = combined_forecast_df.pivot_table(
        index=["Bank", "Best Model"],
        columns="Forecast Quarter",
        values="Forecast ($M)",
        aggfunc="first"
    ).reset_index()

    print("\n" + "=" * 100)
    print("PIVOT FORECAST TABLE | BMO + CIBC")
    print("=" * 100)
    display(forecast_pivot_df)
else:
    print("⚠️ No forecast tables were produced.")

# ==============================================================================
# 9) OPTIONAL: COMBINED SHAP TABLE
# ==============================================================================

if len(all_bank_shap) > 0:
    combined_shap_df = pd.concat(all_bank_shap.values(), axis=0).reset_index(drop=True)
    print("\n" + "=" * 100)
    print("COMBINED SHAP FEATURE IMPORTANCE")
    print("=" * 100)
    display(combined_shap_df)
else:
    print("⚠️ No SHAP outputs were produced.")
```
