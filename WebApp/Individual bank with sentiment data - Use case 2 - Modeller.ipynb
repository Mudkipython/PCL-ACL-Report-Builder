```python
# ==============================================================================
# MODEL 1 (EARNINGS DAY) - PER-BANK FULL PIPELINE
# Updated:
# - OOS evaluation kept
# - Best model selected by LOWEST OOS RMSE for each bank
# - If best model is ML -> use SHAP
# - If best model is NOT ML -> use normal model-based importance for that winner
#   * ARIMAX / VARMAX: abs(exog coefficients)
#   * VAR: abs(lag coefficients into PCL_t+1)
#   * Baseline / SARIMA-blind: no feature importance available
# - Added next 4 quarter forecast using best overall model by LOWEST OOS RMSE
# - Added combined forecast summary + Excel export tabs for forecasts
# - Fcst_Unemp_A_for_t+1 retained
# ==============================================================================

import warnings
warnings.filterwarnings("ignore")

import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
from sklearn.linear_model import ElasticNetCV, RidgeCV, LassoCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.arima.model import ARIMA

# Optional packages
try:
    import shap
    HAS_SHAP = True
except ImportError:
    HAS_SHAP = False

try:
    from pmdarima import auto_arima
    HAS_PMDARIMA = True
except ImportError:
    HAS_PMDARIMA = False

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

try:
    from lightgbm import LGBMRegressor
    HAS_LGBM = True
except ImportError:
    HAS_LGBM = False

PD_VER = int(pd.__version__.split(".")[0])
Q_FREQ = "QE" if PD_VER >= 2 else "Q"

print(">>> Building feature pool for Model 1 (Per-Bank Earnings Day Forecasts)...")
if not HAS_XGB:
    print("⚠️ xgboost not found. Run: pip install xgboost")
if not HAS_LGBM:
    print("⚠️ lightgbm not found. Run: pip install lightgbm")
if not HAS_SHAP:
    print("⚠️ shap not found. Run: pip install shap")

# ==============================================================================
# HELPERS
# ==============================================================================

def clean_num(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    s = str(x).strip().replace(",", "").replace("$", "").replace(" ", "")
    if s in ["—", "", "-", "NA", "N/A", "nan", "NaN", "Unnamed:23", "Unnamed:39"]:
        return np.nan
    if s.startswith("(") and s.endswith(")"):
        s = "-" + s[1:-1]
    try:
        return float(s)
    except:
        return np.nan

def safe_mape(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    denom = np.where(np.abs(y_true) < 1e-8, np.nan, np.abs(y_true))
    return np.nanmean(np.abs((y_true - y_pred) / denom)) * 100

def safe_oos_r2(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    mask = np.isfinite(y_true) & np.isfinite(y_pred)
    y_true = y_true[mask]
    y_pred = y_pred[mask]
    if len(y_true) < 2:
        return np.nan
    try:
        return r2_score(y_true, y_pred)
    except:
        return np.nan

def sanitize_filename(s):
    bad = ['\\', '/', ':', '*', '?', '"', '<', '>', '|']
    out = str(s)
    for ch in bad:
        out = out.replace(ch, "_")
    return out

def is_tree_model(model):
    tree_types = (
        RandomForestRegressor,
        GradientBoostingRegressor,
    )
    if isinstance(model, tree_types):
        return True
    if HAS_XGB and isinstance(model, XGBRegressor):
        return True
    if HAS_LGBM and isinstance(model, LGBMRegressor):
        return True
    return False

def is_linear_model(model):
    linear_types = (
        LassoCV,
        RidgeCV,
        ElasticNetCV,
    )
    return isinstance(model, linear_types)

def is_ml_model_name(model_name):
    ml_prefixes = [
        "06. ML: Lasso",
        "07. ML: Ridge",
        "08. ML: Random Forest",
        "09. ML: Gradient Boosting",
        "10. ML: Elastic Net",
        "11. ML: XGBoost",
        "12. ML: LightGBM",
    ]
    return model_name in ml_prefixes

def is_non_ml_model_name(model_name):
    return model_name in [
        "01. Baseline: Naive (Last Qtr)",
        "02. TS: SARIMA (Blind)",
        "03. TS: VAR (Blind)",
        "04. TS: ARIMAX (Macro-Aware)",
        "05. TS: VARMAX (Macro-Aware)",
    ]

def build_ml_model_dict(tscv):
    ml_models_dict = {
        "06. ML: Lasso": LassoCV(alphas=np.logspace(-2, 4, 50), cv=tscv),
        "07. ML: Ridge": RidgeCV(alphas=np.logspace(-2, 4, 50)),
        "08. ML: Random Forest": RandomForestRegressor(
            n_estimators=100, max_depth=3, random_state=42
        ),
        "09. ML: Gradient Boosting": GradientBoostingRegressor(
            n_estimators=100, random_state=42
        ),
        "10. ML: Elastic Net": ElasticNetCV(
            l1_ratio=[0.1, 0.5, 0.9], cv=tscv, random_state=42
        )
    }

    if HAS_XGB:
        ml_models_dict["11. ML: XGBoost"] = XGBRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42
        )

    if HAS_LGBM:
        ml_models_dict["12. ML: LightGBM"] = LGBMRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42,
            verbose=-1
        )

    return ml_models_dict

def fit_ml_model_by_name(model_name, X_train, y_train, tscv):
    ml_models_dict = build_ml_model_dict(tscv)

    if model_name not in ml_models_dict:
        return None, None, None, None

    model = ml_models_dict[model_name]

    if model_name in ["06. ML: Lasso", "07. ML: Ridge", "10. ML: Elastic Net"]:
        scaler = StandardScaler()
        X_train_fit = scaler.fit_transform(X_train)
        fitted_model = model.fit(X_train_fit, y_train)
        return fitted_model, X_train_fit, list(X_train.columns), scaler
    else:
        fitted_model = model.fit(X_train, y_train)
        return fitted_model, X_train.copy(), list(X_train.columns), None

def save_importance_bar_chart(
    imp_df,
    value_col,
    bank,
    model_name,
    method_label,
    out_dir
):
    if imp_df is None or imp_df.empty:
        return None

    plot_df = imp_df.sort_values(value_col, ascending=True)

    plt.figure(figsize=(10, 6))
    plt.barh(plot_df["Feature"], plot_df[value_col])
    plt.title(f"{method_label} Feature Importance - {bank}\n{model_name}")
    plt.xlabel(value_col)
    plt.ylabel("Feature")
    plt.tight_layout()

    os.makedirs(out_dir, exist_ok=True)
    fname = os.path.join(
        out_dir,
        f"{sanitize_filename(bank)}__{sanitize_filename(model_name)}__{sanitize_filename(method_label)}_bar.png"
    )
    plt.savefig(fname, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close()

    return fname

# ------------------------------------------------------------------------------
# ML feature importance (SHAP preferred)
# ------------------------------------------------------------------------------

def compute_native_ml_feature_importance(model, feature_names):
    try:
        if hasattr(model, "feature_importances_"):
            vals = np.array(model.feature_importances_, dtype=float)
        elif hasattr(model, "coef_"):
            vals = np.array(model.coef_, dtype=float)
            if vals.ndim > 1:
                vals = np.ravel(vals)
            vals = np.abs(vals)
        else:
            return None

        return pd.DataFrame({
            "Feature": feature_names,
            "Importance": vals
        }).sort_values("Importance", ascending=False).reset_index(drop=True)
    except:
        return None

def make_ml_feature_importance(
    model,
    X_used,
    feature_names,
    bank,
    model_name,
    out_dir
):
    if X_used is None or len(feature_names) == 0:
        print(f"⚠️ ML feature importance skipped for {bank}: empty feature matrix.")
        return None, None

    if HAS_SHAP:
        try:
            X_df = pd.DataFrame(X_used, columns=feature_names)

            if is_tree_model(model):
                explainer = shap.TreeExplainer(model)
                shap_values = explainer.shap_values(X_df)
            elif is_linear_model(model):
                explainer = shap.LinearExplainer(model, X_df)
                shap_values = explainer.shap_values(X_df)
            else:
                explainer = shap.Explainer(model.predict, X_df)
                shap_values = explainer(X_df)
                if hasattr(shap_values, "values"):
                    shap_values = shap_values.values

            if isinstance(shap_values, list):
                shap_values = shap_values[0]

            mean_abs_shap = np.abs(shap_values).mean(axis=0)
            imp_df = pd.DataFrame({
                "Feature": feature_names,
                "Importance": mean_abs_shap
            }).sort_values("Importance", ascending=False).reset_index(drop=True)

            save_importance_bar_chart(
                imp_df=imp_df,
                value_col="Importance",
                bank=bank,
                model_name=model_name,
                method_label="SHAP",
                out_dir=out_dir
            )
            return imp_df, "SHAP"
        except Exception as e:
            print(f"⚠️ SHAP failed for {bank} / {model_name}: {e}")

    imp_df = compute_native_ml_feature_importance(model, feature_names)
    if imp_df is not None and not imp_df.empty:
        save_importance_bar_chart(
            imp_df=imp_df,
            value_col="Importance",
            bank=bank,
            model_name=model_name,
            method_label="Native_ML",
            out_dir=out_dir
        )
        return imp_df, "Native_ML"

    return None, None

# ------------------------------------------------------------------------------
# Non-ML feature importance
# ------------------------------------------------------------------------------

def fit_non_ml_winner_and_importance(
    model_name,
    train_valid_df1,
    selected_features_m1,
    exog_m1,
    features_ts_hist
):
    """
    Fit the WINNING NON-ML model on the full bank training data and extract a
    normal feature-importance style table where possible.

    Returns:
        fitted_model,
        imp_df,
        importance_method
    """
    y_full = pd.to_numeric(train_valid_df1["PCL_t+1"], errors="coerce").dropna().copy()

    # 01. Baseline: no explanatory feature importance
    if model_name == "01. Baseline: Naive (Last Qtr)":
        return None, None, "No_Feature_Importance"

    # 02. SARIMA (Blind): no exogenous regressors, so no feature importance
    elif model_name == "02. TS: SARIMA (Blind)":
        try:
            if HAS_PMDARIMA:
                fitted = auto_arima(
                    y_full,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
            else:
                fitted = ARIMA(y_full.values, order=(1, 0, 0)).fit()
            return fitted, None, "No_Feature_Importance"
        except:
            return None, None, "No_Feature_Importance"

    # 03. VAR (Blind): importance from lag-1 coefficients into PCL_t+1
    elif model_name == "03. TS: VAR (Blind)":
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            var_cols = list(dict.fromkeys(var_cols))
            var_input = train_valid_df1[var_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(var_input) <= 5 or var_input.shape[1] < 2:
                return None, None, "VAR_No_Importance"

            fitted = VAR(var_input).fit(maxlags=1)

            params = fitted.params.copy()

            # columns like: const, L1.PCL_t+1, L1.GDP_YoY_t, L1.Overnight_Rate_t
            # index are regressors, columns are dependent variables
            target_col = "PCL_t+1"
            if target_col not in params.columns:
                return fitted, None, "VAR_No_Importance"

            coef_series = params[target_col].copy()
            keep = []
            vals = []

            for idx_name, coef in coef_series.items():
                if str(idx_name).startswith("L1."):
                    feat = str(idx_name).replace("L1.", "")
                    if feat != "PCL_t+1":
                        keep.append(feat)
                        vals.append(abs(float(coef)))

            if len(keep) == 0:
                return fitted, None, "VAR_No_Importance"

            imp_df = pd.DataFrame({
                "Feature": keep,
                "Importance": vals
            }).sort_values("Importance", ascending=False).reset_index(drop=True)

            return fitted, imp_df, "VAR_LagCoef"

        except Exception as e:
            print(f"⚠️ VAR importance failed: {e}")
            return None, None, "VAR_No_Importance"

    # 04. ARIMAX: importance from absolute exogenous coefficients
    elif model_name == "04. TS: ARIMAX (Macro-Aware)":
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            needed_cols = list(dict.fromkeys(needed_cols))
            arimax_train = train_valid_df1[needed_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(arimax_train) <= 5 or len(exog_m1) == 0:
                return None, None, "ARIMAX_No_Importance"

            fitted = ARIMA(
                arimax_train["PCL_t+1"].values,
                exog=arimax_train[exog_m1].values,
                order=(1, 0, 0)
            ).fit()

            param_names = list(fitted.param_names)
            param_vals = np.array(fitted.params, dtype=float)

            rows = []
            for name, val in zip(param_names, param_vals):
                if name in exog_m1:
                    rows.append((name, abs(val)))

            if len(rows) == 0:
                # fallback: align exog names manually if statsmodels renamed things
                exog_count = len(exog_m1)
                if len(param_vals) >= exog_count:
                    maybe_vals = param_vals[:exog_count]
                    rows = list(zip(exog_m1, np.abs(maybe_vals)))

            if len(rows) == 0:
                return fitted, None, "ARIMAX_No_Importance"

            imp_df = pd.DataFrame(rows, columns=["Feature", "Importance"]) \
                .sort_values("Importance", ascending=False).reset_index(drop=True)

            return fitted, imp_df, "ARIMAX_AbsCoef"

        except Exception as e:
            print(f"⚠️ ARIMAX importance failed: {e}")
            return None, None, "ARIMAX_No_Importance"

    # 05. VARMAX: importance from absolute exogenous coefficients on PCL_t+1 equation
    elif model_name == "05. TS: VARMAX (Macro-Aware)":
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            vmax_endog_cols = list(dict.fromkeys(vmax_endog_cols))

            vmax_endog = train_valid_df1[vmax_endog_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(vmax_endog) <= 5 or len(exog_m1) == 0:
                return None, None, "VARMAX_No_Importance"

            vmax_exog = train_valid_df1.loc[vmax_endog.index, exog_m1].apply(pd.to_numeric, errors="coerce").copy()
            valid_idx = vmax_endog.index.intersection(vmax_exog.dropna().index)
            vmax_endog = vmax_endog.loc[valid_idx]
            vmax_exog = vmax_exog.loc[valid_idx]

            if len(vmax_endog) <= 5:
                return None, None, "VARMAX_No_Importance"

            fitted = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)

            param_names = list(fitted.param_names)
            param_vals = np.array(fitted.params, dtype=float)

            rows = []
            for name, val in zip(param_names, param_vals):
                low = str(name).lower()

                # keep exog terms tied to PCL_t+1 equation when identifiable
                if any(ex.lower() in low for ex in exog_m1):
                    if "pcl_t+1" in low or "pcl" in low:
                        matched_feat = None
                        for ex in exog_m1:
                            if ex.lower() in low:
                                matched_feat = ex
                                break
                        if matched_feat is not None:
                            rows.append((matched_feat, abs(float(val))))

            # fallback: aggregate any exog coefficient mention by feature name
            if len(rows) == 0:
                agg = {}
                for name, val in zip(param_names, param_vals):
                    low = str(name).lower()
                    for ex in exog_m1:
                        if ex.lower() in low:
                            agg[ex] = agg.get(ex, 0.0) + abs(float(val))
                rows = list(agg.items())

            if len(rows) == 0:
                return fitted, None, "VARMAX_No_Importance"

            imp_df = pd.DataFrame(rows, columns=["Feature", "Importance"]) \
                .groupby("Feature", as_index=False)["Importance"].sum() \
                .sort_values("Importance", ascending=False).reset_index(drop=True)

            return fitted, imp_df, "VARMAX_AbsCoef"

        except Exception as e:
            print(f"⚠️ VARMAX importance failed: {e}")
            return None, None, "VARMAX_No_Importance"

    return None, None, "No_Feature_Importance"

def make_non_ml_feature_importance(
    model_name,
    train_valid_df1,
    selected_features_m1,
    exog_m1,
    features_ts_hist,
    bank,
    out_dir
):
    fitted_model, imp_df, method_used = fit_non_ml_winner_and_importance(
        model_name=model_name,
        train_valid_df1=train_valid_df1,
        selected_features_m1=selected_features_m1,
        exog_m1=exog_m1,
        features_ts_hist=features_ts_hist
    )

    if imp_df is not None and not imp_df.empty:
        save_importance_bar_chart(
            imp_df=imp_df,
            value_col="Importance",
            bank=bank,
            model_name=model_name,
            method_label=method_used,
            out_dir=out_dir
        )

    return fitted_model, imp_df, method_used

# ------------------------------------------------------------------------------
# Forecast helpers
# ------------------------------------------------------------------------------

def make_future_quarter_index(last_date, steps=4):
    return pd.date_range(
        start=last_date + pd.offsets.QuarterEnd(1),
        periods=steps,
        freq="Q"
    )

def quarter_label(dt):
    return f"{dt.year}Q{dt.quarter}"

def build_future_feature_frame(hist_df, future_index, feature_cols):
    feature_cols = [c for c in list(dict.fromkeys(feature_cols)) if c in hist_df.columns]
    out = pd.DataFrame(index=future_index, columns=feature_cols, dtype=float)

    for col in feature_cols:
        hist_series = pd.to_numeric(hist_df[col], errors="coerce").dropna() if col in hist_df.columns else pd.Series(dtype=float)
        for dt in future_index:
            if dt in hist_df.index and col in hist_df.columns and pd.notna(hist_df.loc[dt, col]):
                out.loc[dt, col] = hist_df.loc[dt, col]
            else:
                out.loc[dt, col] = hist_series.iloc[-1] if len(hist_series) > 0 else np.nan

    out = out.apply(pd.to_numeric, errors="coerce").ffill().bfill()
    return out

def forecast_best_model_next_4q(
    bank,
    best_model_name,
    train_valid_df1,
    bank_df_full,
    selected_features_m1,
    exog_m1,
    features_ts_hist,
    horizon=4
):
    last_train_date = train_valid_df1.index.max()
    future_index = make_future_quarter_index(last_train_date, steps=horizon)
    future_labels = [quarter_label(d) for d in future_index]

    y_full = pd.to_numeric(train_valid_df1["PCL_t+1"], errors="coerce").dropna().copy()

    def make_output(preds):
        return pd.DataFrame({
            "Bank": bank,
            "Forecast Quarter": future_labels,
            "Best Model (Lowest RMSE)": best_model_name,
            "Forecast ($M)": np.array(preds, dtype=float)
        })

    if len(y_full) == 0:
        return make_output([np.nan] * horizon)

    last_level = float(y_full.iloc[-1])

    # 01. Baseline
    if best_model_name == "01. Baseline: Naive (Last Qtr)":
        if "PCL_t" in bank_df_full.columns and pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().shape[0] > 0:
            last_pcl = pd.to_numeric(bank_df_full["PCL_t"], errors="coerce").dropna().iloc[-1]
            return make_output([last_pcl] * horizon)
        return make_output([last_level] * horizon)

    # 02. SARIMA / ARIMA
    elif best_model_name == "02. TS: SARIMA (Blind)":
        try:
            if HAS_PMDARIMA:
                model = auto_arima(
                    y_full,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
                preds = model.predict(n_periods=horizon)
            else:
                model = ARIMA(y_full.values, order=(1, 0, 0)).fit()
                preds = model.forecast(steps=horizon)
            return make_output(preds)
        except:
            return make_output([last_level] * horizon)

    # 03. VAR
    elif best_model_name == "03. TS: VAR (Blind)":
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            var_cols = list(dict.fromkeys(var_cols))
            var_input = train_valid_df1[var_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_fit = VAR(var_input).fit(maxlags=1)
                current_values = var_input.values.copy()
                preds = []

                last_macro = current_values[-1, 1:] if current_values.shape[1] > 1 else np.array([])

                for _ in range(horizon):
                    fc = var_fit.forecast(current_values[-var_fit.k_ar:], steps=1)[0]
                    if len(last_macro) > 0:
                        fc[1:] = last_macro
                    preds.append(fc[0])
                    current_values = np.vstack([current_values, fc])

                return make_output(preds)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 04. ARIMAX
    elif best_model_name == "04. TS: ARIMAX (Macro-Aware)":
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            needed_cols = list(dict.fromkeys(needed_cols))
            arimax_train = train_valid_df1[needed_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(arimax_train) > 5 and len(exog_m1) > 0:
                future_exog = build_future_feature_frame(bank_df_full, future_index, exog_m1)[exog_m1]

                model = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=arimax_train[exog_m1].values,
                    order=(1, 0, 0)
                ).fit()

                preds = model.forecast(steps=horizon, exog=future_exog.values)
                return make_output(preds)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 05. VARMAX
    elif best_model_name == "05. TS: VARMAX (Macro-Aware)":
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_valid_df1.columns]
            vmax_endog_cols = list(dict.fromkeys(vmax_endog_cols))

            vmax_endog = train_valid_df1[vmax_endog_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_valid_df1.loc[vmax_endog.index, exog_m1].apply(pd.to_numeric, errors="coerce").copy()
                valid_idx = vmax_endog.index.intersection(vmax_exog.dropna().index)
                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]

                future_exog = build_future_feature_frame(bank_df_full, future_index, exog_m1)[exog_m1]

                model = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)
                fc = model.forecast(steps=horizon, exog=future_exog)
                preds = fc.iloc[:, 0].values
                return make_output(preds)
            else:
                return make_output([last_level] * horizon)
        except:
            return make_output([last_level] * horizon)

    # 06+ ML
    elif "ML:" in best_model_name:
        try:
            ml_cols = selected_features_m1 + ["PCL_t+1"]
            final_ml_df = train_valid_df1[ml_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()

            if len(final_ml_df) < 6 or len(selected_features_m1) == 0:
                return make_output([last_level] * horizon)

            n_splits_final = min(3, len(final_ml_df) - 1)
            if n_splits_final < 2:
                return make_output([last_level] * horizon)

            tscv_final = TimeSeriesSplit(n_splits=n_splits_final)
            ml_models_final = build_ml_model_dict(tscv_final)

            if best_model_name not in ml_models_final:
                return make_output([last_level] * horizon)

            X_train = final_ml_df[selected_features_m1].copy()
            y_train = final_ml_df["PCL_t+1"].values

            model = ml_models_final[best_model_name]

            is_linear = best_model_name in ["06. ML: Lasso", "07. ML: Ridge", "10. ML: Elastic Net"]

            if is_linear:
                scaler = StandardScaler()
                X_train_fit = scaler.fit_transform(X_train)
                fitted_model = model.fit(X_train_fit, y_train)
            else:
                scaler = None
                fitted_model = model.fit(X_train, y_train)

            future_features = build_future_feature_frame(bank_df_full, future_index, selected_features_m1).copy()

            preds = []
            last_pred = y_train[-1]

            for i in range(horizon):
                row = future_features.iloc[i:i+1].copy()

                if "PCL_t" in row.columns:
                    row.loc[:, "PCL_t"] = last_pred

                row = row[selected_features_m1].astype(float)

                if scaler is not None:
                    row_fit = scaler.transform(row)
                else:
                    row_fit = row

                pred = float(fitted_model.predict(row_fit)[0])
                preds.append(pred)
                last_pred = pred

            return make_output(preds)

        except Exception as e:
            print(f"⚠️ Forecasting failed for {bank} / {best_model_name}: {e}")
            return make_output([last_level] * horizon)

    else:
        return make_output([last_level] * horizon)

# ==============================================================================
# LOAD DATA
# ==============================================================================

file_hist = "Complete_UC2_Forecast_Data.xlsx"
raw_df = pd.read_excel(file_hist, sheet_name="Forecast Data", header=0)

raw_df.columns = [str(c).strip() for c in raw_df.columns]

print("\nDetected columns:")
print(raw_df.columns.tolist())

required_base_cols = ["Bank", "Year", "Quarter"]
missing_base_cols = [c for c in required_base_cols if c not in raw_df.columns]
if missing_base_cols:
    raise ValueError(f"Missing required columns: {missing_base_cols}")

cols_to_clean = [
    "PCL_t+1",
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1",
    "GDP_YoY_t",
    "Overnight_Rate_t"
]

for c in cols_to_clean:
    if c in raw_df.columns:
        raw_df[c] = raw_df[c].apply(clean_num)

raw_df["Bank"] = raw_df["Bank"].astype(str).str.strip()
raw_df["Year"] = pd.to_numeric(raw_df["Year"], errors="coerce")
raw_df["Quarter"] = raw_df["Quarter"].astype(str).str.strip()

raw_df = raw_df.dropna(subset=["Bank", "Year", "Quarter"])
raw_df = raw_df.sort_values(by=["Bank", "Year", "Quarter"]).reset_index(drop=True)

cols_to_fill = [
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

for c in cols_to_fill:
    if c in raw_df.columns:
        raw_df[c] = raw_df.groupby("Bank")[c].transform(lambda x: x.ffill().bfill())

q_map = {"Q1": "03-31", "Q2": "06-30", "Q3": "09-30", "Q4": "12-31"}

candidate_features_m1 = [
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

features_ts_hist = ["GDP_YoY_t", "Overnight_Rate_t"]

# ==============================================================================
# BUILD PER-BANK DATASETS
# ==============================================================================

bank_model_data = {}

for bank in sorted(raw_df["Bank"].dropna().unique()):
    bank_df = raw_df[raw_df["Bank"] == bank].copy()

    bank_df["date_idx"] = pd.to_datetime(
        bank_df["Year"].astype(int).astype(str) + "-" + bank_df["Quarter"].map(q_map),
        errors="coerce"
    )

    bank_df = bank_df.dropna(subset=["date_idx"]).set_index("date_idx").sort_index()
    bank_df = bank_df[bank_df.index <= "2026-03-31"]

    keep_cols = ["Bank", "PCL_t+1", "Year", "Quarter"] + candidate_features_m1 + features_ts_hist
    keep_cols = [c for c in keep_cols if c in bank_df.columns]

    model1_df_bank = bank_df[keep_cols].copy()
    bank_model_data[bank] = model1_df_bank

print(f"\n✅ Setup complete. Built Model 1 datasets for {len(bank_model_data)} banks.\n")
for bank, df_ in bank_model_data.items():
    if "PCL_t" in df_.columns and df_["PCL_t"].dropna().shape[0] > 0:
        last_pcl = df_["PCL_t"].dropna().iloc[-1]
        print(f"{bank}: Latest PCL_t = ${last_pcl:,.0f}M")
    else:
        print(f"{bank}: Latest PCL_t = NaN")

# ==============================================================================
# MODEL BACKTEST PER BANK
# ==============================================================================

print("\n>>> Executing Model 1 Pipeline PER BANK...\n")

all_bank_results = {}
all_bank_scorecards = {}
bank_winner_summary = []
all_feature_importance = {}
all_bank_forecasts = {}

importance_output_dir = "feature_importance_bank"

for bank, model1_df in bank_model_data.items():
    print("\n" + "=" * 100)
    print(f"🏦 BANK: {bank}")
    print("=" * 100)

    train_valid_df1 = model1_df.dropna(subset=["PCL_t+1"]).copy()

    if len(train_valid_df1) < 16:
        print(f"⚠️ Skipping {bank}: not enough history for stable backtesting.")
        continue

    # --------------------------------------------------------------------------
    # 1) FEATURE SELECTION
    # --------------------------------------------------------------------------
    fs_feature_pool = [c for c in candidate_features_m1 if c in train_valid_df1.columns]
    fs_df = train_valid_df1.dropna(subset=fs_feature_pool + ["PCL_t+1"]).copy()

    if len(fs_df) < 10:
        selected_features_m1 = [
            c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
            if c in train_valid_df1.columns
        ]
    else:
        n_splits_fs = min(3, len(fs_df) - 1)

        if n_splits_fs < 2:
            selected_features_m1 = [
                c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
                if c in train_valid_df1.columns
            ]
        else:
            tscv_fs = TimeSeriesSplit(n_splits=n_splits_fs)
            X_cand_sc1 = StandardScaler().fit_transform(fs_df[fs_feature_pool])
            y_full1 = fs_df["PCL_t+1"].values

            elastic_fs1 = ElasticNetCV(
                l1_ratio=[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1.0],
                cv=tscv_fs,
                random_state=42
            )
            elastic_fs1.fit(X_cand_sc1, y_full1)

            selected_features_m1 = []
            print(f"\n🔍 FEATURE SELECTION: {bank}")
            print("-" * 85)
            for feat, coef in zip(fs_feature_pool, elastic_fs1.coef_):
                status = "✅ KEPT" if abs(coef) > 1e-4 else "❌ PRUNED"
                if abs(coef) > 1e-4:
                    selected_features_m1.append(feat)
                print(f"  {status:10s} : {feat:35s} (Weight: {coef:>8.2f})")

            if len(selected_features_m1) <= 1:
                selected_features_m1 = [
                    c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
                    if c in train_valid_df1.columns
                ]

    exog_m1 = [f for f in selected_features_m1 if f != "PCL_t"]
    if not exog_m1:
        exog_m1 = [
            c for c in ["Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
            if c in train_valid_df1.columns
        ]

    # --------------------------------------------------------------------------
    # 2) MODEL SETUP
    # --------------------------------------------------------------------------
    n_splits_bt = min(3, len(train_valid_df1) - 1)
    if n_splits_bt < 2:
        print(f"⚠️ Skipping {bank}: insufficient rows for TimeSeriesSplit.")
        continue

    tscv = TimeSeriesSplit(n_splits=n_splits_bt)
    ml_models_dict = build_ml_model_dict(tscv)

    h_test = min(8, max(1, len(train_valid_df1) // 3))
    test_indices = list(range(len(train_valid_df1) - h_test, len(train_valid_df1)))

    prediction_records1 = []

    # --------------------------------------------------------------------------
    # 3) WALK-FORWARD BACKTEST
    # --------------------------------------------------------------------------
    for t in test_indices:
        if t < 12:
            continue

        train_df = train_valid_df1.iloc[:t].copy()
        test_row = train_valid_df1.iloc[t:t+1].copy()

        y_tr_abs = train_df["PCL_t+1"]
        y_actual = float(test_row["PCL_t+1"].iloc[0])

        forecast_date = test_row.index[0]
        yr = int(forecast_date.year)
        qtr = f"Q{forecast_date.quarter}"

        def eval_mod(m_name, p_val):
            prediction_records1.append({
                "Bank": bank,
                "Year": yr,
                "Quarter": qtr,
                "Model": m_name,
                "Predicted ($M)": float(np.squeeze(p_val)) if pd.notna(p_val) else np.nan,
                "Actual ($M)": y_actual
            })

        # 01. Baseline
        try:
            baseline_pred = train_df["PCL_t"].iloc[-1]
            eval_mod("01. Baseline: Naive (Last Qtr)", baseline_pred)
        except:
            eval_mod("01. Baseline: Naive (Last Qtr)", np.nan)

        # 02. SARIMA / ARIMA
        try:
            if HAS_PMDARIMA:
                sarima_model = auto_arima(
                    y_tr_abs,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
                sar_p = sarima_model.predict(n_periods=1)[0]
            else:
                sarima_model = ARIMA(y_tr_abs.values, order=(1, 0, 0)).fit()
                sar_p = sarima_model.forecast(steps=1)[0]
            eval_mod("02. TS: SARIMA (Blind)", sar_p)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("02. TS: SARIMA (Blind)", fallback)

        # 03. VAR
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            var_input = train_df[var_cols].dropna()
            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_fit = VAR(var_input).fit(maxlags=1)
                var_p = var_fit.forecast(var_input.values[-1:], steps=1)[0, 0]
                eval_mod("03. TS: VAR (Blind)", var_p)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("03. TS: VAR (Blind)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("03. TS: VAR (Blind)", fallback)

        # 04. ARIMAX
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            arimax_train = train_df[needed_cols].dropna()
            arimax_test = test_row[exog_m1].copy()

            if len(arimax_train) > 5 and not arimax_test.isna().any(axis=None):
                arimax_mod = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=arimax_train[exog_m1].values,
                    order=(1, 0, 0)
                ).fit()

                arimax_fcst = arimax_mod.forecast(steps=1, exog=arimax_test[exog_m1].values)[0]
                eval_mod("04. TS: ARIMAX (Macro-Aware)", arimax_fcst)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)

        # 05. VARMAX
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            vmax_endog = train_df[vmax_endog_cols].dropna()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_df.loc[vmax_endog.index, exog_m1]
                valid_idx = vmax_exog.dropna().index.intersection(vmax_endog.index)
                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]
                test_exog = test_row[exog_m1]

                if len(vmax_endog) > 5 and not test_exog.isna().any(axis=None):
                    vmax_mod = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(
                        disp=False, maxiter=200
                    )
                    vmax_fcst = vmax_mod.forecast(steps=1, exog=test_exog).iloc[0, 0]
                    eval_mod("05. TS: VARMAX (Macro-Aware)", vmax_fcst)
                else:
                    fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                    eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)

        # 06+ ML models
        ml_cols = selected_features_m1 + ["PCL_t+1"]
        ml_train = train_df[ml_cols].dropna().copy() if len(selected_features_m1) > 0 else pd.DataFrame()
        ml_test = test_row[selected_features_m1].copy() if len(selected_features_m1) > 0 else pd.DataFrame()

        if len(ml_train) > 5 and not ml_test.isna().any(axis=None):
            X_tr_raw = ml_train[selected_features_m1]
            y_tr_ml = ml_train["PCL_t+1"].values
            X_te_raw = ml_test[selected_features_m1]

            for m_name, model in ml_models_dict.items():
                try:
                    if m_name in ["06. ML: Lasso", "07. ML: Ridge", "10. ML: Elastic Net"]:
                        sc = StandardScaler()
                        X_tr = sc.fit_transform(X_tr_raw)
                        X_te = sc.transform(X_te_raw)
                    else:
                        X_tr = X_tr_raw
                        X_te = X_te_raw

                    fitted_model = model.fit(X_tr, y_tr_ml)
                    pred = fitted_model.predict(X_te)[0]
                    eval_mod(m_name, pred)
                except:
                    eval_mod(m_name, np.nan)
        else:
            for m_name in ml_models_dict.keys():
                eval_mod(m_name, np.nan)

    # --------------------------------------------------------------------------
    # 4) EVALUATION
    # --------------------------------------------------------------------------
    results_df1 = pd.DataFrame(prediction_records1)

    if results_df1.empty:
        print(f"⚠️ No valid OOS predictions for {bank}.")
        continue

    metrics_list1 = []
    for model_name, grp in results_df1.groupby("Model"):
        valid = grp.dropna(subset=["Predicted ($M)", "Actual ($M)"]).copy()

        if len(valid) > 0:
            y_a = valid["Actual ($M)"].values.astype(float)
            y_p = valid["Predicted ($M)"].values.astype(float)

            metrics_list1.append({
                "Bank": bank,
                "Model": model_name,
                "OOS R2": float(safe_oos_r2(y_a, y_p)),
                "OOS MAPE": float(safe_mape(y_a, y_p)),
                "OOS RMSE": float(np.sqrt(np.mean((y_a - y_p) ** 2))),
                "N_OOS": len(valid)
            })

    final_table1 = pd.DataFrame(metrics_list1).sort_values(
        by=["OOS RMSE", "OOS MAPE"]
    ).reset_index(drop=True)

    all_bank_results[bank] = results_df1
    all_bank_scorecards[bank] = final_table1

    print(f"\n📊 SCORECARD: {bank}")
    display(final_table1)

    # best overall by lowest OOS RMSE
    best_row = final_table1.iloc[0]
    best_model_name = best_row["Model"]

    # choose importance logic based on actual winner
    if is_ml_model_name(best_model_name):
        importance_model_name = best_model_name
        importance_basis = "Winner is ML -> SHAP"
    else:
        importance_model_name = best_model_name
        importance_basis = "Winner is non-ML -> normal model-based importance"

    bank_winner_summary.append({
        "Bank": bank,
        "Winning Model": best_row["Model"],
        "Best OOS RMSE": best_row["OOS RMSE"],
        "Best OOS MAPE": best_row["OOS MAPE"],
        "Best OOS R2": best_row["OOS R2"],
        "OOS Points": best_row["N_OOS"],
        "Feature Importance Model Used": importance_model_name,
        "Feature Importance Basis": importance_basis
    })

    print(f"🏆 Best overall model for {bank}: {best_row['Model']} (selected by lowest OOS RMSE)")
    print(f"🔍 Feature importance model for {bank}: {importance_model_name} ({importance_basis})")

    # --------------------------------------------------------------------------
    # 5) FIT FINAL IMPORTANCE MODEL ON FULL BANK DATA
    # --------------------------------------------------------------------------
    try:
        if is_ml_model_name(importance_model_name):
            final_ml_df = train_valid_df1[selected_features_m1 + ["PCL_t+1"]].dropna().copy()

            if len(final_ml_df) >= 6 and len(selected_features_m1) > 0:
                X_final = final_ml_df[selected_features_m1].copy()
                y_final = final_ml_df["PCL_t+1"].values

                n_splits_final = min(3, len(final_ml_df) - 1)
                if n_splits_final >= 2:
                    tscv_final = TimeSeriesSplit(n_splits=n_splits_final)

                    fitted_model, X_for_importance, feature_names, scaler = fit_ml_model_by_name(
                        importance_model_name,
                        X_final,
                        y_final,
                        tscv_final
                    )

                    if fitted_model is not None:
                        imp_df, method_used = make_ml_feature_importance(
                            model=fitted_model,
                            X_used=X_for_importance,
                            feature_names=feature_names,
                            bank=bank,
                            model_name=importance_model_name,
                            out_dir=importance_output_dir
                        )

                        if imp_df is not None:
                            imp_df = imp_df.copy()
                            imp_df["Bank"] = bank
                            imp_df["Importance Model"] = importance_model_name
                            imp_df["Importance Basis"] = importance_basis
                            imp_df["Importance Method"] = method_used
                            all_feature_importance[bank] = imp_df

                            print(f"\nTop feature importance for {bank}:")
                            display(imp_df.sort_values("Importance", ascending=False).reset_index(drop=True))
                    else:
                        print(f"⚠️ Could not fit ML importance model for {bank}.")
                else:
                    print(f"⚠️ Not enough rows to fit final ML importance model for {bank}.")
            else:
                print(f"⚠️ Not enough clean ML data for importance for {bank}.")

        else:
            fitted_model, imp_df, method_used = make_non_ml_feature_importance(
                model_name=importance_model_name,
                train_valid_df1=train_valid_df1,
                selected_features_m1=selected_features_m1,
                exog_m1=exog_m1,
                features_ts_hist=features_ts_hist,
                bank=bank,
                out_dir=importance_output_dir
            )

            if imp_df is not None:
                imp_df = imp_df.copy()
                imp_df["Bank"] = bank
                imp_df["Importance Model"] = importance_model_name
                imp_df["Importance Basis"] = importance_basis
                imp_df["Importance Method"] = method_used
                all_feature_importance[bank] = imp_df

                print(f"\nTop feature importance for {bank}:")
                display(imp_df.sort_values("Importance", ascending=False).reset_index(drop=True))
            else:
                print(f"⚠️ No meaningful feature importance available for {bank} / {importance_model_name}.")

    except Exception as e:
        print(f"⚠️ Final feature-importance block failed for {bank}: {e}")

    # --------------------------------------------------------------------------
    # 6) NEXT 4 QUARTER FORECAST USING BEST MODEL BY RMSE
    # --------------------------------------------------------------------------
    try:
        forecast_df = forecast_best_model_next_4q(
            bank=bank,
            best_model_name=best_model_name,
            train_valid_df1=train_valid_df1,
            bank_df_full=model1_df,
            selected_features_m1=selected_features_m1,
            exog_m1=exog_m1,
            features_ts_hist=features_ts_hist,
            horizon=4
        )

        all_bank_forecasts[bank] = forecast_df

        print(f"\n📈 NEXT 4 QUARTER FORECAST FOR {bank} USING BEST MODEL: {best_model_name}")
        print("-" * 100)
        display(forecast_df)

    except Exception as e:
        print(f"⚠️ Forecast block failed for {bank}: {e}")

# ==============================================================================
# 7) COMBINED WINNER SUMMARY
# ==============================================================================

if len(bank_winner_summary) > 0:
    summary_df = pd.DataFrame(bank_winner_summary).sort_values(
        by=["Best OOS RMSE", "Best OOS MAPE"]
    ).reset_index(drop=True)

    print("\n" + "=" * 100)
    print("FINAL BANK WINNER SUMMARY")
    print("=" * 100)
    display(summary_df)
else:
    print("⚠️ No bank-level summaries were produced.")

# ==============================================================================
# 8) COMBINED FORECAST SUMMARY
# ==============================================================================

if len(all_bank_forecasts) > 0:
    combined_forecast_df = pd.concat(all_bank_forecasts.values(), axis=0).reset_index(drop=True)

    print("\n" + "=" * 100)
    print("COMBINED NEXT 4 QUARTER FORECAST SUMMARY (BEST MODEL BY RMSE)")
    print("=" * 100)
    display(combined_forecast_df)

    combined_forecast_pivot = combined_forecast_df.pivot_table(
        index=["Bank", "Best Model (Lowest RMSE)"],
        columns="Forecast Quarter",
        values="Forecast ($M)",
        aggfunc="first"
    ).reset_index()

    print("\n" + "=" * 100)
    print("PIVOT TABLE: NEXT 4 QUARTER FORECAST (BEST MODEL BY RMSE)")
    print("=" * 100)
    display(combined_forecast_pivot)
else:
    print("⚠️ No forecasts were produced.")

# ==============================================================================
# 9) OPTIONAL: EXPORT RESULTS
# ==============================================================================

export_to_excel = False

if export_to_excel and len(all_bank_scorecards) > 0:
    out_file = "model1_per_bank_oos_results_with_winner_based_importance_and_forecast_rmse.xlsx"
    with pd.ExcelWriter(out_file, engine="openpyxl") as writer:
        if len(bank_winner_summary) > 0:
            summary_df.to_excel(writer, sheet_name="Winner_Summary", index=False)

        if len(all_bank_forecasts) > 0:
            combined_forecast_df.to_excel(writer, sheet_name="Forecast_Combined", index=False)
            combined_forecast_pivot.to_excel(writer, sheet_name="Forecast_Pivot", index=False)

        for bank in all_bank_scorecards.keys():
            bank_safe = sanitize_filename(bank)[:25]

            scorecard_df = all_bank_scorecards.get(bank, pd.DataFrame())
            preds_df = all_bank_results.get(bank, pd.DataFrame())
            imp_df = all_feature_importance.get(bank, pd.DataFrame())
            fcst_df = all_bank_forecasts.get(bank, pd.DataFrame())

            if not scorecard_df.empty:
                scorecard_df.to_excel(writer, sheet_name=f"{bank_safe}_Score", index=False)

            if not preds_df.empty:
                preds_df.to_excel(writer, sheet_name=f"{bank_safe}_Pred", index=False)

            if not imp_df.empty:
                imp_df.sort_values("Importance", ascending=False).to_excel(
                    writer,
                    sheet_name=f"{bank_safe}_Imp",
                    index=False
                )

            if not fcst_df.empty:
                fcst_df.to_excel(writer, sheet_name=f"{bank_safe}_Fcst", index=False)

    print(f"\n✅ Exported results to: {out_file}")

print(f"\n✅ Feature importance charts saved in folder: {importance_output_dir}")
```

```python
# ==============================================================================
# MODEL 1 (EARNINGS DAY) - PER-BANK FULL PIPELINE
# No total aggregation; runs separately for each bank
# Updated:
# - interaction features removed
# - Train R2 -> OOS R2
# - Fcst_Unemp_Q_for_t+1 -> Fcst_Unemp_A_for_t+1
# - keep data from 2021Q4 onward only
# - fixed OOS horizon to 4
# - add SHAP feature importance on best ML model (lowest OOS MAPE) per bank
# - SHAP feature importance shown with bar chart
# ==============================================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.linear_model import ElasticNetCV, RidgeCV, LassoCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import TimeSeriesSplit

# Optional packages
try:
    from pmdarima import auto_arima
    HAS_PMDARIMA = True
except ImportError:
    HAS_PMDARIMA = False

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

try:
    HAS_XGB = HAS_XGB
except:
    HAS_XGB = False

try:
    from lightgbm import LGBMRegressor
    HAS_LGBM = True
except ImportError:
    HAS_LGBM = False

try:
    import shap
    HAS_SHAP = True
except ImportError:
    HAS_SHAP = False

PD_VER = int(pd.__version__.split(".")[0])
Q_FREQ = "QE" if PD_VER >= 2 else "Q"

print(">>> Building feature pool for Model 1 (Per-Bank Earnings Day Forecasts)...")
if not HAS_XGB or not HAS_LGBM:
    print("⚠️ Notice: xgboost or lightgbm not found. Run '!pip install xgboost lightgbm' if needed.")
if not HAS_SHAP:
    print("⚠️ Notice: shap not found. Run '!pip install shap' if needed.")

# ==============================================================================
# HELPERS
# ==============================================================================

def clean_num(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    s = str(x).strip().replace(",", "").replace("$", "").replace(" ", "")
    if s in ["—", "", "-", "NA", "N/A", "nan", "NaN", "Unnamed:23", "Unnamed:39"]:
        return np.nan
    if s.startswith("(") and s.endswith(")"):
        s = "-" + s[1:-1]
    try:
        return float(s)
    except:
        return np.nan

def safe_mape(y_true, y_pred):
    y_true = np.array(y_true, dtype=float)
    y_pred = np.array(y_pred, dtype=float)
    denom = np.where(np.abs(y_true) < 1e-8, np.nan, np.abs(y_true))
    return np.nanmean(np.abs((y_true - y_pred) / denom)) * 100

def plot_shap_bar(shap_df, bank, model_name):
    shap_plot_df = shap_df.sort_values(by="SHAP Importance", ascending=True)

    plt.figure(figsize=(8, max(4, 0.6 * len(shap_plot_df))))
    plt.barh(shap_plot_df["Feature"], shap_plot_df["SHAP Importance"])
    plt.xlabel("Mean |SHAP value|")
    plt.ylabel("Feature")
    plt.title(f"SHAP Feature Importance - {bank}\n{model_name}")
    plt.tight_layout()
    plt.show()

# ==============================================================================
# LOAD DATA
# ==============================================================================

file_hist = "Complete_UC2_Forecast_Data.xlsx"

# IMPORTANT: use header=0, not header=1
raw_df = pd.read_excel(file_hist, sheet_name="Forecast Data", header=0)

# Clean column names
raw_df.columns = [str(c).strip() for c in raw_df.columns]

print("\nDetected columns:")
print(raw_df.columns.tolist())

required_base_cols = ["Bank", "Year", "Quarter"]
missing_base_cols = [c for c in required_base_cols if c not in raw_df.columns]
if missing_base_cols:
    raise ValueError(f"Missing required columns: {missing_base_cols}")

cols_to_clean = [
    "PCL_t+1",
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1",
    "GDP_YoY_t",
    "Overnight_Rate_t"
]

for c in cols_to_clean:
    if c in raw_df.columns:
        raw_df[c] = raw_df[c].apply(clean_num)

raw_df["Bank"] = raw_df["Bank"].astype(str).str.strip()
raw_df["Year"] = pd.to_numeric(raw_df["Year"], errors="coerce")
raw_df["Quarter"] = raw_df["Quarter"].astype(str).str.strip()

raw_df = raw_df.dropna(subset=["Bank", "Year", "Quarter"])
raw_df = raw_df.sort_values(by=["Bank", "Year", "Quarter"]).reset_index(drop=True)

cols_to_fill = [
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

for c in cols_to_fill:
    if c in raw_df.columns:
        raw_df[c] = raw_df.groupby("Bank")[c].transform(lambda x: x.ffill().bfill())

q_map = {"Q1": "03-31", "Q2": "06-30", "Q3": "09-30", "Q4": "12-31"}

candidate_features_m1 = [
    "PCL_t",
    "Sent_BART_t",
    "Fcst_GDP_YoY_for_t+1",
    "Fcst_Unemp_A_for_t+1",
    "Fcst_BoC_for_t+1"
]

features_ts_hist = ["GDP_YoY_t", "Overnight_Rate_t"]

# ==============================================================================
# BUILD PER-BANK DATASETS
# ==============================================================================

bank_model_data = {}

for bank in sorted(raw_df["Bank"].dropna().unique()):
    bank_df = raw_df[raw_df["Bank"] == bank].copy()

    bank_df["date_idx"] = pd.to_datetime(
        bank_df["Year"].astype(int).astype(str) + "-" + bank_df["Quarter"].map(q_map),
        errors="coerce"
    )

    bank_df = bank_df.dropna(subset=["date_idx"]).set_index("date_idx").sort_index()

    # Keep data from 2021Q4 onward only
    bank_df = bank_df[bank_df.index >= "2021-12-31"]

    # Optional upper cutoff retained from your earlier code
    bank_df = bank_df[bank_df.index <= "2026-03-31"]

    keep_cols = ["Bank", "PCL_t+1", "Year", "Quarter"] + candidate_features_m1 + features_ts_hist
    keep_cols = [c for c in keep_cols if c in bank_df.columns]

    model1_df_bank = bank_df[keep_cols].copy()
    bank_model_data[bank] = model1_df_bank

print(f"\n✅ Setup complete. Built Model 1 datasets for {len(bank_model_data)} banks.\n")
for bank, df_ in bank_model_data.items():
    if "PCL_t" in df_.columns and df_["PCL_t"].dropna().shape[0] > 0:
        last_pcl = df_["PCL_t"].dropna().iloc[-1]
        print(f"{bank}: Latest PCL_t = ${last_pcl:,.0f}M")
    else:
        print(f"{bank}: Latest PCL_t = NaN")

# ==============================================================================
# MODEL BACKTEST PER BANK
# ==============================================================================

print("\n>>> Executing Model 1 Pipeline PER BANK...\n")

all_bank_results = {}
all_bank_scorecards = {}
bank_winner_summary = []

for bank, model1_df in bank_model_data.items():
    print("\n" + "=" * 100)
    print(f"🏦 BANK: {bank}")
    print("=" * 100)

    train_valid_df1 = model1_df.dropna(subset=["PCL_t+1"]).copy()

    if len(train_valid_df1) < 10:
        print(f"⚠️ Skipping {bank}: not enough history after 2021Q4 filter.")
        continue

    # --------------------------------------------------------------------------
    # 1) FEATURE SELECTION
    # --------------------------------------------------------------------------
    fs_feature_pool = [c for c in candidate_features_m1 if c in train_valid_df1.columns]
    fs_df = train_valid_df1.dropna(subset=fs_feature_pool + ["PCL_t+1"]).copy()

    if len(fs_df) < 6:
        selected_features_m1 = [
            c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
            if c in train_valid_df1.columns
        ]
    else:
        n_splits_fs = min(3, len(fs_df) - 1)
        if n_splits_fs < 2:
            selected_features_m1 = [
                c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
                if c in train_valid_df1.columns
            ]
        else:
            tscv_fs = TimeSeriesSplit(n_splits=n_splits_fs)
            X_cand_sc1 = StandardScaler().fit_transform(fs_df[fs_feature_pool])
            y_full1 = fs_df["PCL_t+1"].values

            elastic_fs1 = ElasticNetCV(
                l1_ratio=[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1.0],
                cv=tscv_fs,
                random_state=42
            )
            elastic_fs1.fit(X_cand_sc1, y_full1)

            selected_features_m1 = []
            print(f"\n🔍 FEATURE SELECTION: {bank}")
            print("-" * 85)
            for feat, coef in zip(fs_feature_pool, elastic_fs1.coef_):
                status = "✅ KEPT" if abs(coef) > 1e-4 else "❌ PRUNED"
                if abs(coef) > 1e-4:
                    selected_features_m1.append(feat)
                print(f"  {status:10s} : {feat:35s} (Weight: {coef:>8.2f})")

            if len(selected_features_m1) <= 1:
                selected_features_m1 = [
                    c for c in ["PCL_t", "Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
                    if c in train_valid_df1.columns
                ]

    exog_m1 = [f for f in selected_features_m1 if f != "PCL_t"]
    if not exog_m1:
        exog_m1 = [
            c for c in ["Fcst_GDP_YoY_for_t+1", "Sent_BART_t"]
            if c in train_valid_df1.columns
        ]

    # --------------------------------------------------------------------------
    # 2) MODEL SETUP
    # --------------------------------------------------------------------------
    n_splits_bt = min(3, len(train_valid_df1) - 1)
    if n_splits_bt < 2:
        print(f"⚠️ Skipping {bank}: insufficient rows for TimeSeriesSplit.")
        continue

    tscv = TimeSeriesSplit(n_splits=n_splits_bt)

    ml_models_dict = {
        "06. ML: Lasso": LassoCV(alphas=np.logspace(-2, 4, 50), cv=tscv),
        "07. ML: Ridge": RidgeCV(alphas=np.logspace(-2, 4, 50)),
        "08. ML: Random Forest": RandomForestRegressor(n_estimators=100, max_depth=3, random_state=42),
        "09. ML: Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
        "10. ML: Elastic Net": ElasticNetCV(l1_ratio=[0.1, 0.5, 0.9], cv=tscv, random_state=42)
    }

    if HAS_XGB:
        ml_models_dict["11. ML: XGBoost"] = XGBRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42
        )

    if HAS_LGBM:
        ml_models_dict["12. ML: LightGBM"] = LGBMRegressor(
            n_estimators=100,
            max_depth=3,
            learning_rate=0.05,
            random_state=42,
            verbose=-1
        )

    # Fixed OOS horizon = 4
    h_test = 4
    if len(train_valid_df1) <= h_test:
        print(f"⚠️ Skipping {bank}: not enough rows to reserve 4 OOS periods.")
        continue

    test_indices = list(range(len(train_valid_df1) - h_test, len(train_valid_df1)))
    prediction_records1 = []

    # --------------------------------------------------------------------------
    # 3) WALK-FORWARD BACKTEST
    # --------------------------------------------------------------------------
    for t in test_indices:
        if t < 4:
            continue

        train_df = train_valid_df1.iloc[:t].copy()
        test_row = train_valid_df1.iloc[t:t+1].copy()

        y_tr_abs = train_df["PCL_t+1"]
        y_actual = float(test_row["PCL_t+1"].iloc[0])

        forecast_date = test_row.index[0]
        yr = int(forecast_date.year)
        qtr = f"Q{forecast_date.quarter}"

        def eval_mod(m_name, p_val):
            prediction_records1.append({
                "Bank": bank,
                "Year": yr,
                "Quarter": qtr,
                "Model": m_name,
                "Predicted ($M)": float(np.squeeze(p_val)) if pd.notna(p_val) else np.nan,
                "Actual ($M)": y_actual
            })

        # 01. Baseline
        try:
            baseline_pred = train_df["PCL_t"].iloc[-1]
            eval_mod("01. Baseline: Naive (Last Qtr)", baseline_pred)
        except:
            eval_mod("01. Baseline: Naive (Last Qtr)", np.nan)

        # 02. SARIMA / ARIMA
        try:
            if HAS_PMDARIMA:
                sarima_model = auto_arima(
                    y_tr_abs,
                    seasonal=False,
                    suppress_warnings=True,
                    error_action="ignore"
                )
                sar_p = sarima_model.predict(n_periods=1)[0]
            else:
                sarima_model = ARIMA(y_tr_abs.values, order=(1, 0, 0)).fit()
                sar_p = sarima_model.forecast(steps=1)[0]
            eval_mod("02. TS: SARIMA (Blind)", sar_p)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("02. TS: SARIMA (Blind)", fallback)

        # 03. VAR
        try:
            var_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            var_input = train_df[var_cols].dropna()
            if len(var_input) > 5 and var_input.shape[1] >= 2:
                var_fit = VAR(var_input).fit(maxlags=1)
                var_p = var_fit.forecast(var_input.values[-1:], steps=1)[0, 0]
                eval_mod("03. TS: VAR (Blind)", var_p)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("03. TS: VAR (Blind)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("03. TS: VAR (Blind)", fallback)

        # 04. ARIMAX
        try:
            needed_cols = ["PCL_t+1"] + exog_m1
            arimax_train = train_df[needed_cols].dropna()
            arimax_test = test_row[exog_m1].copy()

            if len(arimax_train) > 5 and not arimax_test.isna().any(axis=None):
                arimax_mod = ARIMA(
                    arimax_train["PCL_t+1"].values,
                    exog=arimax_train[exog_m1].values,
                    order=(1, 0, 0)
                ).fit()

                arimax_fcst = arimax_mod.forecast(steps=1, exog=arimax_test[exog_m1].values)[0]
                eval_mod("04. TS: ARIMAX (Macro-Aware)", arimax_fcst)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("04. TS: ARIMAX (Macro-Aware)", fallback)

        # 05. VARMAX
        try:
            vmax_endog_cols = ["PCL_t+1"] + [c for c in features_ts_hist if c in train_df.columns]
            vmax_endog = train_df[vmax_endog_cols].dropna()

            if len(vmax_endog) > 5 and len(exog_m1) > 0:
                vmax_exog = train_df.loc[vmax_endog.index, exog_m1]
                valid_idx = vmax_exog.dropna().index.intersection(vmax_endog.index)
                vmax_endog = vmax_endog.loc[valid_idx]
                vmax_exog = vmax_exog.loc[valid_idx]
                test_exog = test_row[exog_m1]

                if len(vmax_endog) > 5 and not test_exog.isna().any(axis=None):
                    vmax_mod = VARMAX(vmax_endog, exog=vmax_exog, order=(1, 0)).fit(disp=False, maxiter=200)
                    vmax_fcst = vmax_mod.forecast(steps=1, exog=test_exog).iloc[0, 0]
                    eval_mod("05. TS: VARMAX (Macro-Aware)", vmax_fcst)
                else:
                    fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                    eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
            else:
                fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
                eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)
        except:
            fallback = train_df["PCL_t"].iloc[-1] if "PCL_t" in train_df.columns else np.nan
            eval_mod("05. TS: VARMAX (Macro-Aware)", fallback)

        # 06+ ML models
        ml_cols = selected_features_m1 + ["PCL_t+1"]
        ml_train = train_df[ml_cols].dropna().copy() if len(selected_features_m1) > 0 else pd.DataFrame()
        ml_test = test_row[selected_features_m1].copy() if len(selected_features_m1) > 0 else pd.DataFrame()

        if len(ml_train) > 5 and not ml_test.isna().any(axis=None):
            sc = StandardScaler()
            X_tr_sc = sc.fit_transform(ml_train[selected_features_m1])
            y_tr_ml = ml_train["PCL_t+1"].values
            X_te_sc = sc.transform(ml_test[selected_features_m1])

            for m_name, model in ml_models_dict.items():
                try:
                    fitted_model = model.fit(X_tr_sc, y_tr_ml)
                    pred = fitted_model.predict(X_te_sc)[0]
                    eval_mod(m_name, pred)
                except:
                    eval_mod(m_name, np.nan)
        else:
            for m_name in ml_models_dict.keys():
                eval_mod(m_name, np.nan)

    # --------------------------------------------------------------------------
    # 4) EVALUATION
    # --------------------------------------------------------------------------
    results_df1 = pd.DataFrame(prediction_records1)

    if results_df1.empty:
        print(f"⚠️ No valid OOS predictions for {bank}.")
        continue

    metrics_list1 = []
    for model_name, grp in results_df1.groupby("Model"):
        valid = grp.dropna(subset=["Predicted ($M)", "Actual ($M)"]).copy()

        if len(valid) > 0:
            y_a = valid["Actual ($M)"].values.astype(float)
            y_p = valid["Predicted ($M)"].values.astype(float)

            oos_r2 = r2_score(y_a, y_p) if len(y_a) > 1 else np.nan

            metrics_list1.append({
                "Bank": bank,
                "Model": model_name,
                "OOS R2": float(oos_r2) if pd.notna(oos_r2) else np.nan,
                "OOS MAPE": float(safe_mape(y_a, y_p)),
                "OOS RMSE": float(np.sqrt(np.mean((y_a - y_p) ** 2))),
                "N_OOS": len(valid)
            })

    final_table1 = pd.DataFrame(metrics_list1).sort_values(by="OOS MAPE").reset_index(drop=True)

    all_bank_results[bank] = results_df1
    all_bank_scorecards[bank] = final_table1

    display(final_table1)

    best_row = final_table1.iloc[0]
    ml_only = final_table1[final_table1["Model"].str.contains("ML:", na=False)].copy()
    best_ml_m1 = ml_only.iloc[0]["Model"] if not ml_only.empty else None

    bank_winner_summary.append({
        "Bank": bank,
        "Winning Model": best_row["Model"],
        "Best OOS MAPE": best_row["OOS MAPE"],
        "Best OOS RMSE": best_row["OOS RMSE"],
        "Best OOS R2": best_row["OOS R2"],
        "OOS Points": best_row["N_OOS"],
        "Best ML Model": best_ml_m1
    })

    print(f"🏆 Best overall model for {bank}: {best_row['Model']}")
    if best_ml_m1 is not None:
        print(f"🚀 Best ML model for {bank}: {best_ml_m1}")

    # --------------------------------------------------------------------------
    # 5) SHAP FEATURE IMPORTANCE FOR BEST ML MODEL
    # --------------------------------------------------------------------------
    if best_ml_m1 is not None:
        print(f"\n📊 SHAP FEATURE IMPORTANCE: {bank} | {best_ml_m1}")
        print("-" * 100)

        if not HAS_SHAP:
            print("⚠️ SHAP package not installed. Run: !pip install shap")
        else:
            ml_cols = selected_features_m1 + ["PCL_t+1"]
            full_ml_df = train_valid_df1[ml_cols].dropna().copy()

            if len(full_ml_df) > 5 and len(selected_features_m1) > 0:
                X_full = full_ml_df[selected_features_m1].copy()
                y_full = full_ml_df["PCL_t+1"].values

                sc_full = StandardScaler()
                X_full_sc = sc_full.fit_transform(X_full)
                X_full_sc_df = pd.DataFrame(X_full_sc, columns=selected_features_m1, index=X_full.index)

                best_model_obj = ml_models_dict.get(best_ml_m1, None)

                if best_model_obj is None:
                    print(f"⚠️ Could not locate best ML model object for {bank}.")
                else:
                    try:
                        best_model_fit = best_model_obj.fit(X_full_sc, y_full)

                        if any(k in best_ml_m1 for k in ["Random Forest", "Gradient Boosting", "XGBoost", "LightGBM"]):
                            explainer = shap.TreeExplainer(best_model_fit)
                            shap_values = explainer.shap_values(X_full_sc_df)

                            if isinstance(shap_values, list):
                                shap_values = shap_values[0]
                        else:
                            explainer = shap.Explainer(best_model_fit.predict, X_full_sc_df)
                            shap_exp = explainer(X_full_sc_df)
                            shap_values = shap_exp.values

                        shap_values = np.array(shap_values)

                        if shap_values.ndim == 3:
                            shap_values = shap_values[:, :, 0]

                        shap_importance = np.abs(shap_values).mean(axis=0)

                        shap_df = pd.DataFrame({
                            "Feature": selected_features_m1,
                            "SHAP Importance": shap_importance
                        }).sort_values(by="SHAP Importance", ascending=False).reset_index(drop=True)

                        print(shap_df.to_string(index=False))
                        display(shap_df)

                        plot_shap_bar(shap_df, bank, best_ml_m1)

                    except Exception as e:
                        print(f"⚠️ SHAP failed for {bank}: {e}")
            else:
                print(f"⚠️ Not enough valid data to calculate SHAP for {bank}.")

# ==============================================================================
# 6) COMBINED WINNER SUMMARY
# ==============================================================================

if len(bank_winner_summary) > 0:
    summary_df = pd.DataFrame(bank_winner_summary).sort_values(by="Best OOS MAPE").reset_index(drop=True)

    print("\n" + "=" * 100)
    print("FINAL BANK WINNER SUMMARY")
    print("=" * 100)
    display(summary_df)
else:
    print("⚠️ No bank-level summaries were produced.")
```
